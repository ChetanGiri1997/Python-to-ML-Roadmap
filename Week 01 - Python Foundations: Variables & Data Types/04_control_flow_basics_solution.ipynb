21.Find max in a list using loop.

```python
numbers = [3, 7, 2, 9, 5]

max_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num

print(max_value)
```

22.Check palindrome using loop.

```python
text = "madam"
is_palindrome = True

for i in range(len(text) // 2):
    if text[i] != text[len(text) - 1 - i]:
        is_palindrome = False
        break

if is_palindrome:
    print("Palindrome")
else:
    print("Not Palindrome")
```

23.Generate Fibonacci with loop.

```python
n = 10

a, b = 0, 1

for _ in range(n):
    print(a)
    a, b = b, a + b
```
