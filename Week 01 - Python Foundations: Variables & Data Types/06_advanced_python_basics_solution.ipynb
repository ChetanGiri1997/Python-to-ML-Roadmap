8.Simulate a queue using a list and loops (enqueue/dequeue operations).

```python
queue = []

# Enqueue operation
for i in range(5):
    queue.append(i)
    print("Enqueued:", i, "Queue:", queue)

print("\n--- Dequeue Operations ---\n")

# Dequeue operation
while len(queue) > 0:
    removed = queue.pop(0)
    print("Dequeued:", removed, "Queue:", queue)
```

9.Simulate a stack using a list and loops (push/pop operations).

10.Use a loop to compute the transpose of a matrix.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

# Create empty transpose matrix
transpose = []

# Loop through columns of original matrix
for j in range(len(matrix[0])):
    new_row = []
    
    # Loop through rows
    for i in range(len(matrix)):
        new_row.append(matrix[i][j])
    
    transpose.append(new_row)

# Print result
print("Original Matrix:")
for row in matrix:
    print(row)

print("\nTranspose Matrix:")
for row in transpose:
    print(row)
```

12.Check if two strings are anagrams using loops.

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Remove spaces and convert to lowercase for fair comparison
str1 = str1.replace(" ", "").lower()
str2 = str2.replace(" ", "").lower()

is_anagram = True

if len(str1) != len(str2):
    is_anagram = False
else:
    for char in str1:
        if str1.count(char) != str2.count(char):
            is_anagram = False
            break

if is_anagram:
    print("Strings are anagrams")
else:
    print("Strings are NOT anagrams")
```

13.Write a program to check if a sentence is a pangram (contains all alphabets).

```python
import string

sentence = input("Enter a sentence: ")

# Convert to lowercase for uniform comparison
sentence = sentence.lower()

alphabet = "abcdefghijklmnopqrstuvwxyz"

is_pangram = True

for char in alphabet:
    if char not in sentence:
        is_pangram = False
        break

if is_pangram:
    print("The sentence is a pangram")
else:
    print("The sentence is NOT a pangram")
```

14.Implement a simple Caesar cipher using loops.

```python
text = input("Enter text: ")
shift = int(input("Enter shift value: "))

result = ""

for char in text:
    if char.isalpha():
        # Handle lowercase
        if char.islower():
            base = ord('a')
            new_char = chr((ord(char) - base + shift) % 26 + base)
        # Handle uppercase
        else:
            base = ord('A')
            new_char = chr((ord(char) - base + shift) % 26 + base)

        result += new_char
    else:
        # Keep spaces, punctuation unchanged
        result += char

print("Encrypted text:", result)
```

15.Generate all permutations of a small string using loops (no recursion).

```python
string = input("Enter a small string: ")

permutations = [""]

for char in string:
    new_perms = []
    
    for p in permutations:
        for i in range(len(p) + 1):
            new_perms.append(p[:i] + char + p[i:])
    
    permutations = new_perms

print("All permutations:")
for p in permutations:
    print(p)
```

17.Calculate variance of a list manually using loops.

```python
numbers = [10, 20, 30, 40, 50]

# Step 1: Calculate mean
total = 0
n = 0

for num in numbers:
    total += num
    n += 1

mean = total / n

# Step 2: Calculate squared differences
sq_diff_sum = 0

for num in numbers:
    sq_diff = (num - mean) ** 2
    sq_diff_sum += sq_diff

# Step 3: Variance
variance = sq_diff_sum / n

print("Numbers:", numbers)
print("Mean:", mean)
print("Variance:", variance)
```

18.Implement Bubble Sort using loops.

```python
numbers = [64, 34, 25, 12, 22, 11, 90]

n = len(numbers)

for i in range(n):
    swapped = False

    for j in range(0, n - i - 1):
        if numbers[j] > numbers[j + 1]:
            # swap
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
            swapped = True

    # Optimization: stop if no swaps occurred
    if not swapped:
        break

print("Sorted list:", numbers)
```

19.Implement Selection Sort using loops.

```python
numbers = [64, 25, 12, 22, 11]

n = len(numbers)

for i in range(n):
    min_index = i

    # Find the minimum element in remaining unsorted part
    for j in range(i + 1, n):
        if numbers[j] < numbers[min_index]:
            min_index = j

    # Swap the found minimum with first element of unsorted part
    numbers[i], numbers[min_index] = numbers[min_index], numbers[i]

print("Sorted list:", numbers)
```

20.Implement Insertion Sort using loops.

```python
numbers = [64, 25, 12, 22, 11]

for i in range(1, len(numbers)):
    key = numbers[i]
    j = i - 1

    # Shift elements of sorted part that are greater than key
    while j >= 0 and numbers[j] > key:
        numbers[j + 1] = numbers[j]
        j -= 1

    # Insert key at correct position
    numbers[j + 1] = key

print("Sorted list:", numbers)
```

22.Implement Binary Search manually using loops.

```python
numbers = [10, 20, 30, 40, 50, 60, 70, 80]  # must be sorted
target = int(input("Enter number to search: "))

low = 0
high = len(numbers) - 1

found = False

while low <= high:
    mid = (low + high) // 2

    if numbers[mid] == target:
        print("Element found at index:", mid)
        found = True
        break
    elif numbers[mid] < target:
        low = mid + 1
    else:
        high = mid - 1

if not found:
    print("Element not found")
```
