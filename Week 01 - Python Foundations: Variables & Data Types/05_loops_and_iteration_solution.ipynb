12.Find duplicates in a list using loops.

```python
numbers = [1, 2, 3, 2, 4, 5, 1, 6]

duplicates = []

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] == numbers[j] and numbers[i] not in duplicates:
            duplicates.append(numbers[i])

print(duplicates)
```

13.Write a loop to check if a number is prime.

```python
num = 29

is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print("Prime number")
else:
    print("Not a prime number")
```

17.Implement a manual sorting algorithm (loop-based, no built-in sort).

```python
numbers = [5, 2, 9, 1, 6]

n = len(numbers)

for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

print(numbers)
```

20.Find the GCD (Greatest Common Divisor) of two numbers using a loop.

```python
a = 48
b = 18

gcd = 1

for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print(gcd)
```

21.Find the LCM (Least Common Multiple) of two numbers using a loop.

```python
a = 12
b = 18

lcm = max(a, b)

while True:
    if lcm % a == 0 and lcm % b == 0:
        break
    lcm += 1

print(lcm)
```

22.Check if a number is an Armstrong number using a loop.

```python
num = 153

temp = num
digits = len(str(num))
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** digits
    temp //= 10

if total == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

23.Convert a decimal number to binary using a loop.

```python
num = 13
binary = ""

while num > 0:
    remainder = num % 2
    binary = str(remainder) + binary
    num = num // 2

print(binary)
```

24.Convert a decimal number to hexadecimal using a loop.

```python
num = 254
hex_digits = "0123456789ABCDEF"
result = ""

while num > 0:
    remainder = num % 16
    result = hex_digits[remainder] + result
    num = num // 16

print(result)
```
