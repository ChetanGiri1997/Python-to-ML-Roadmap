### Q1. Writing to a Text File
Write a function `write_lines(filename, lines)` that writes a list of strings to a file, one per line, using a `with open(...)` context manager. Test it by writing 5 lines to `sample.txt`.

```python
def write_lines(filename, lines):
    with open(filename, "w") as file:
        for line in lines:
            file.write(line + "\n")


# Test the function
lines = [
    "First line",
    "Second line",
    "Third line",
    "Fourth line",
    "Fifth line"
]

write_lines("sample.txt", lines)

print("5 lines written to sample.txt")
```

### Q5. Checking If a File Exists Before Reading
Using the `os` or `pathlib` module, write a function `safe_read(filename)` that checks whether a file exists before attempting to read it, returning an appropriate message if it does not.

```python
from pathlib import Path

def safe_read(filename):
    path = Path(filename)

    if path.exists():
        with open(filename, "r") as file:
            return file.read()
    else:
        return f"Error: '{filename}' does not exist."


# Test cases
print(safe_read("sample.txt"))      # Existing file
print(safe_read("missing.txt"))     # Non-existent file
```

Alternative Solution Using os

```python
import os

def safe_read(filename):
    if os.path.exists(filename):
        with open(filename, "r") as file:
            return file.read()
    else:
        return f"Error: '{filename}' does not exist."
```

```python
safe_read("missing.txt")
```

### Q7. Catching Multiple Exception Types
Write a function `safe_divide(a, b)` that handles both `ZeroDivisionError` (when `b` is 0) and `TypeError` (when `a` or `b` is not a number), printing a different friendly message for each case.
 

```python
def safe_divide(a, b):
    try:
        result = a / b
        return result

    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")

    except TypeError:
        print("Error: Both inputs must be numbers.")
```

```python
print(safe_divide(10, 2))    # valid case
print(safe_divide(10, 0))    # ZeroDivisionError
print(safe_divide(10, "a"))  # TypeError
```

Write a function `parse_number(s)` that attempts to convert `s` to a float, prints an error message in the `except` block if it fails, and prints a success message in the `else` block if it succeeds (only runs when no exception occurred).

### Q9. Using `finally`
Write a function `process_file(filename)` that opens a file, does some processing (e.g., counts lines), and uses a `finally` block to print `"File processing attempt complete"` regardless of whether an exception occurred.

```python
def process_file(filename):
    try:
        with open(filename, "r") as file:
            lines = file.readlines()
            print(f"Number of lines: {len(lines)}")

    except FileNotFoundError:
        print("Error: File not found.")

    finally:
        print("File processing attempt complete")
```

```python
process_file("sample.txt")      # existing file case
process_file("missing.txt")     # non-existent file case
```

### Q10. Raising Your Own Exceptions
Write a function `withdraw(balance, amount)` that raises a `ValueError` with a descriptive message if `amount` is greater than `balance`, otherwise returns the new balance after subtraction.

### Q11. Catching and Re-Raising
Write a function `risky_operation(x)` that catches a `ZeroDivisionError` internally, prints a log message, and then re-raises the same exception so the caller is still aware something went wrong.

```python
def risky_operation(x):
    try:
        result = 10 / x
        return result

    except ZeroDivisionError as e:
        print("Log: Attempted division by zero in risky_operation")
        raise   # re-raises the same exception to the caller
```

### Q13. Custom Exception with Extra Data
Extend `InsufficientFundsError` from Q12 to store the `balance` and `amount_requested` as attributes on the exception object, accessible when it's caught (e.g., `except InsufficientFundsError as e: print(e.balance)`).

```python
class InsufficientFundsError(Exception):
    def __init__(self, balance, amount_requested):
        self.balance = balance
        self.amount_requested = amount_requested
        
        message = (
            f"Insufficient funds: attempted to withdraw {amount_requested}, "
            f"but only {balance} available"
        )
        super().__init__(message)
```

```python
def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    return balance - amount
```

```python
try:
    withdraw(500, 800)

except InsufficientFundsError as e:
    print("Error message:", e)
    print("Balance:", e.balance)
    print("Requested:", e.amount_requested)
```

### Q14. Exception Hierarchy
Define a base custom exception `AppError(Exception)` and two subclasses `ValidationError(AppError)` and `NotFoundError(AppError)`. Write a function that raises one or the other based on input, and a single `except AppError` block that catches both, printing which specific type occurred.

```python
class AppError(Exception):
    pass


class ValidationError(AppError):
    pass


class NotFoundError(AppError):
    pass
```

```python
def process_input(value):
    if value == "invalid":
        raise ValidationError("Input failed validation rules")
    elif value == "missing":
        raise NotFoundError("Requested item not found")
    else:
        return f"Processed: {value}"
```

```python
test_values = ["invalid", "missing", "ok"]

for v in test_values:
    try:
        print(process_input(v))

    except AppError as e:
        print(f"AppError occurred: {type(e).__name__} -> {e}")
```

### Q15. Working with CSV Files — Writing
Using the `csv` module, write a function `write_csv(filename, rows)` that writes a list of lists (rows) to a CSV file, including a header row `["Name", "Age", "City"]`.

```python
import csv

def write_csv(filename, rows):
    header = ["Name", "Age", "City"]
    
    with open(filename, "w", newline="") as file:
        writer = csv.writer(file)
        
        # write header first
        writer.writerow(header)
        
        # write data rows
        writer.writerows(rows)
```

```python
data = [
    ["Alice", 25, "Kathmandu"],
    ["Bob", 30, "Lalitpur"],
    ["Charlie", 22, "Bhaktapur"]
]

write_csv("people.csv", data)
print("CSV file written successfully.")
```

### Q17. Working with JSON Files — Writing
Write a function `save_json(filename, data)` that takes a Python dictionary and writes it to a file as JSON using the `json` module, with indentation for readability.

### Q18. Working with JSON Files — Reading
Write a function `load_json(filename)` that reads a JSON file and returns it as a Python dictionary. Handle the case where the file contains invalid JSON by catching `json.JSONDecodeError` and returning `None`.
 

```python
import json

def load_json(filename):
    try:
        with open(filename, "r") as file:
            return json.load(file)

    except json.JSONDecodeError:
        print("Error: File contains invalid JSON.")
        return None

    except FileNotFoundError:
        print("Error: File not found.")
        return None
```

### Q19. Context Manager Behavior Demonstration

```python
def demo_context_manager(filename):
    file = None

    try:
        with open(filename, "w") as file:
            file.write("Testing context manager behavior\n")

            # intentional error
            result = 10 / 0

    except ZeroDivisionError:
        print("Caught division by zero error outside with-block")

    # check if file is closed
    print("File closed:", file.closed if file else "File was never opened")


# Run test
demo_context_manager("test.txt")
```

### Q20. Exception Chaining with `raise ... from ...`
Write a function that catches a low-level exception (e.g., `ValueError` from a failed conversion) and raises a more descriptive custom exception using `raise NewException("message") from original_exception`. Print the result to show the chained traceback relationship.

```python
class DataProcessingError(Exception):
    pass
```

```python
def convert_to_int(value):
    try:
        return int(value)

    except ValueError as e:
        raise DataProcessingError(
            f"Failed to process input: {value}"
        ) from e
```

```python
try:
    convert_to_int("abc")

except DataProcessingError as e:
    print("Custom exception caught:", e)
    print("Original cause:", e.__cause__)
```

### Q21. Validating User Input with a Loop and Exceptions
Write a function `get_positive_integer(prompts)` that simulates input validation: given a list of strings `prompts` representing simulated user inputs (since we can't use real `input()` in a notebook easily, simulate it), it loops through them and returns the first one that successfully converts to a positive integer, using `try`/`except` and `continue`.
 

```python
def get_positive_integer(prompts):
    for value in prompts:
        try:
            num = int(value)

            if num > 0:
                return num
            else:
                continue

        except ValueError:
            continue

    return None
```

```python
inputs = ["abc", "-5", "0", "12", "99"]

result = get_positive_integer(inputs)
print(result)
```

### Q22. Logging Errors to a File
Write a function `safe_operation(func, *args)` that attempts to call `func(*args)`, and if any exception occurs, appends a line describing the error (exception type and message) to an `error_log.txt` file instead of letting the program crash. Test it with a function call that's guaranteed to fail.

```python
def safe_operation(func, *args):
    try:
        return func(*args)

    except Exception as e:
        with open("error_log.txt", "a") as log_file:
            log_file.write(f"{type(e).__name__}: {e}\n")
```

```python
def divide(a, b):
    return a / b
```

```python
def copy_file(source, destination):
    with open(source, "r") as f1, open(destination, "w") as f2:
        for line in f1:
            f2.write(line)
```

### Q24. Custom Exception for Data Validation Pipeline
Define a custom exception `SchemaValidationError`. Write a function `validate_record(record, required_fields)` that checks whether a dictionary `record` contains all keys listed in `required_fields`, raising `SchemaValidationError` with a message listing any missing fields if not.
 

```python
class SchemaValidationError(Exception):
    pass
```

```python
def validate_record(record, required_fields):
    missing = [field for field in required_fields if field not in record]

    if missing:
        raise SchemaValidationError(
            f"Missing required fields: {', '.join(missing)}"
        )

    return True
```

```python
record1 = {"name": "Alice", "age": 25, "city": "Kathmandu"}

print(validate_record(record1, ["name", "age", "city"]))
```

```python
record2 = {"name": "Bob", "age": 30}

try:
    validate_record(record2, ["name", "age", "city"])
except SchemaValidationError as e:
    print("Validation failed:", e)
```

### Q25. Challenge — Robust File-Based Data Processor
Write a function `process_data_file(filename)` that:
1. Tries to open and read a JSON file containing a list of records (each a dictionary with `"name"` and `"score"` keys)
2. Catches `FileNotFoundError` and returns an empty list with a warning message if the file doesn't exist
3. Catches `json.JSONDecodeError` if the file is malformed
4. For each record, validates that `"score"` is a number using a `try`/`except`, skipping (and logging) any invalid records instead of crashing
5. Returns the average score of all valid records, or `0` if there are none
Test it against at least two scenarios: a valid JSON file, and a missing file.

```python
import json

def process_data_file(filename):
    try:
        with open(filename, "r") as file:
            data = json.load(file)

    except FileNotFoundError:
        print("Warning: File not found.")
        return []

    except json.JSONDecodeError:
        print("Error: Malformed JSON file.")
        return 0

    total = 0
    count = 0

    for record in data:
        try:
            score = record["score"]

            # ensure numeric
            score = float(score)

            total += score
            count += 1

        except (KeyError, ValueError, TypeError):
            print(f"Skipping invalid record: {record}")
            continue

    return total / count if count > 0 else 0
```

```python
valid_data = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 90},
    {"name": "Charlie", "score": "95"}
]

with open("valid.json", "w") as f:
    json.dump(valid_data, f)
```

```python
result = process_data_file("valid.json")
print("Average score:", result)
```
