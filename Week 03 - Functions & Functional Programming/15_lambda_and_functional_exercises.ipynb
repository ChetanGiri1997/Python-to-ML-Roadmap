# Day 15 — Lambda Expressions & Functional Programming
 
**Week 3: Functions & Functional Programming**
 
Topics covered: `lambda`, `map()`, `filter()`, `reduce()`, `sorted()` with key functions, list/dict/set comprehensions as functional tools, `itertools` and `functools` basics.

### Q1. Basic Lambda
Write a lambda function that takes a number and returns its square. Assign it to a variable `square` and test it with 3 different inputs.

### Q2. Lambda with Multiple Arguments
Write a lambda function `multiply` that takes two arguments and returns their product. Test it with at least 3 pairs of numbers.


### Q3. Lambda Inside a Function Call (No Variable Assignment)
Use an anonymous lambda directly inside a call to `sorted()` to sort a list of tuples `[(1, 'b'), (3, 'a'), (2, 'c')]` by the second element of each tuple.


### Q4. `map()` with Lambda
Given a list `prices = [100, 250, 99, 400]`, use `map()` with a lambda to create a new list where each price has 10% tax added.

### Q5. `filter()` with Lambda
Given a list `numbers = [12, 7, 23, 4, 9, 18, 33, 2]`, use `filter()` with a lambda to extract only the even numbers.

### Q6. `map()` and `filter()` Combined
Given a list of words, use `filter()` to keep words longer than 4 characters, then use `map()` to convert the remaining words to uppercase. Do this in one chained expression.

### Q7. `functools.reduce()` Basics
Import `reduce` from `functools` and use it with a lambda to compute the product of all numbers in a list `[1, 2, 3, 4, 5]`.

### Q8. `reduce()` to Find Maximum
Use `reduce()` with a lambda to find the maximum value in a list of numbers, without using the built-in `max()` function.

### Q9. Sorting with `key=`
Given a list of dictionaries representing people (`[{"name": "Ram", "age": 25}, {"name": "Sita", "age": 19}, ...]`), use `sorted()` with a `key=lambda` to sort the list by age.

### Q10. Sorting in Descending Order
Using the same list of dictionaries from Q9, sort by age in descending order using `reverse=True`.

### Q11. Sorting by Multiple Keys
Given a list of tuples `[("Asha", 25, "HR"), ("Bibek", 25, "IT"), ("Chandra", 22, "Finance")]`, sort first by age, then by name, using a lambda that returns a tuple `(age, name)`.

### Q12. Lambda with Conditional Expression
Write a lambda `classify = lambda x: "even" if x % 2 == 0 else "odd"`. Use `map()` to apply it to a list of numbers and print the results.

### Q13. List Comprehension vs `map()`/`filter()`
Take the same task from Q5 (filtering even numbers) and rewrite it using a list comprehension instead of `filter()`. Compare both versions in a comment — which do you find more readable?

### Q14. Dictionary Comprehension
Given a list of words, create a dictionary comprehension that maps each word to its length.

### Q15. Set Comprehension
Given a sentence string, use a set comprehension to extract all unique vowels found in the sentence.

### Q16. Nested List Comprehension
Use a nested list comprehension to create a 3x3 multiplication table as a list of lists (each inner list representing a row).

### Q17. `itertools.chain`
Import `chain` from `itertools` and use it to combine three separate lists into a single iterable, then convert it to a list and print it.

### Q18. `itertools.combinations`
Use `itertools.combinations` to generate all possible pairs (combinations of 2) from the list `["A", "B", "C", "D"]`.

### Q19. `itertools.permutations`
Use `itertools.permutations` to generate all possible orderings of the list `[1, 2, 3]`.


### Q20. `itertools.count` with `zip`
Use `itertools.count(start=1)` together with `zip()` to create numbered labels for a list of items (e.g., `["1: apple", "2: banana", "3: cherry"]`), without using `enumerate()`.

### Q21. Lambda for Custom Sort of Strings by Last Character
Given a list of strings, sort them based on their last character using `sorted()` with a lambda key.

### Q22. Functional Pipeline with `map`, `filter`, `reduce`
Given a list of numbers from 1 to 20:
1. Filter out numbers not divisible by 3
2. Map each remaining number to its square
3. Reduce the squared values to a single sum
Do this as a clear sequence of steps (can be chained or stepwise).

### Q23. Comparing Comprehensions vs Functional Style — Word Frequency Prep
Given a sentence, use a dictionary comprehension to build a mapping of each unique word to its length. Then do the equivalent using `map()` and `dict()`/`zip()`. Compare the two approaches in a comment.

### Q24. Lambda Sorting Key with `abs()`
Given a list of numbers including negatives, e.g., `[-10, 3, -1, 7, -8, 2]`, use `sorted()` with `key=lambda x: abs(x)` to sort by absolute value while preserving original signs in the output.

### Q25. Challenge — Functional Data Pipeline
Given a list of dictionaries representing products: `[{"name": "Pen", "price": 20, "qty": 5}, {"name": "Notebook", "price": 60, "qty": 2}, ...]` (create at least 6 entries), write a functional pipeline (using `map`, `filter`, `reduce`, and/or comprehensions, plus lambdas) that:
1. Filters out products with `qty < 1`
2. Computes `total = price * qty` for each remaining product
3. Returns the grand total revenue across all valid products
4. Also returns the name of the single highest-revenue product
Avoid explicit `for` loops — solve it functionally.
