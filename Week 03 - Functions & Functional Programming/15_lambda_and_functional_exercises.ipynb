### Q6. `map()` and `filter()` Combined
Given a list of words, use `filter()` to keep words longer than 4 characters, then use `map()` to convert the remaining words to uppercase. Do this in one chained expression.

```python
words = ["apple", "cat", "banana", "dog", "elephant", "sun", "computer"]

# filter words longer than 4 chars, then convert to uppercase
result = list(map(lambda w: w.upper(),
                  filter(lambda w: len(w) > 4, words)))

print(result)
```

### Q20. `itertools.count` with `zip`
Use `itertools.count(start=1)` together with `zip()` to create numbered labels for a list of items (e.g., `["1: apple", "2: banana", "3: cherry"]`), without using `enumerate()`.

```python
from itertools import count

items = ["apple", "banana", "cherry"]

# Create a counter starting from 1
counter = count(start=1)

# Zip numbers with items
numbered = [f"{i}: {item}" for i, item in zip(counter, items)]

print(numbered)
```

### Q22. Functional Pipeline with `map`, `filter`, `reduce`
Given a list of numbers from 1 to 20:
1. Filter out numbers not divisible by 3
2. Map each remaining number to its square
3. Reduce the squared values to a single sum
Do this as a clear sequence of steps (can be chained or stepwise).

```python
from functools import reduce

numbers = list(range(1, 21))

# Step 1: Filter numbers divisible by 3
filtered = filter(lambda x: x % 3 == 0, numbers)

# Step 2: Square each remaining number
mapped = map(lambda x: x ** 2, filtered)

# Step 3: Reduce to a single sum
result = reduce(lambda a, b: a + b, mapped)

print(result)
```

### Q23. Comparing Comprehensions vs Functional Style — Word Frequency Prep
Given a sentence, use a dictionary comprehension to build a mapping of each unique word to its length. Then do the equivalent using `map()` and `dict()`/`zip()`. Compare the two approaches in a comment.

```python
sentence = "python is powerful and python is simple"

words = sentence.split()

# dictionary comprehension
word_lengths = {word: len(word) for word in set(words)}

print(word_lengths)
```

Functional style (map + zip)

```python
sentence = "python is powerful and python is simple"

words = list(set(sentence.split()))

# map lengths
lengths = map(len, words)

# combine using zip + dict
word_lengths = dict(zip(words, lengths))

print(word_lengths)
```


Comparison:

1. Dictionary Comprehension:
   - Cleaner and more readable
   - Directly expresses intent: word -> length
   - Preferred in modern Python

2. map() + zip() + dict():
   - More functional programming style
   - Breaks logic into steps (less readable)
   - Useful in pipeline-heavy or FP-oriented code

Conclusion:
Dictionary comprehension is more Pythonic and preferred for clarity.
Functional style is more verbose but useful for teaching FP concepts.

### Q25. Challenge — Functional Data Pipeline
Given a list of dictionaries representing products: `[{"name": "Pen", "price": 20, "qty": 5}, {"name": "Notebook", "price": 60, "qty": 2}, ...]` (create at least 6 entries), write a functional pipeline (using `map`, `filter`, `reduce`, and/or comprehensions, plus lambdas) that:
1. Filters out products with `qty < 1`
2. Computes `total = price * qty` for each remaining product
3. Returns the grand total revenue across all valid products
4. Also returns the name of the single highest-revenue product
Avoid explicit `for` loops — solve it functionally.

```python
products = [
    {"name": "Pen", "price": 20, "qty": 5},
    {"name": "Notebook", "price": 60, "qty": 2},
    {"name": "Pencil", "price": 10, "qty": 0},
    {"name": "Bag", "price": 500, "qty": 1},
    {"name": "Eraser", "price": 5, "qty": 10},
    {"name": "Marker", "price": 30, "qty": 3}
]
```

```python
from functools import reduce

# Step 1: Filter valid products (qty >= 1)
valid_products = filter(lambda p: p["qty"] >= 1, products)

# Step 2: Compute revenue for each product
revenues = list(map(lambda p: {
    "name": p["name"],
    "revenue": p["price"] * p["qty"]
}, valid_products))

# Step 3: Grand total revenue
total_revenue = reduce(lambda a, b: a + b,
                       map(lambda x: x["revenue"], revenues))

# Step 4: Highest revenue product
top_product = reduce(lambda a, b: a if a["revenue"] > b["revenue"] else b,
                     revenues)

print("Total Revenue:", total_revenue)
print("Top Product:", top_product["name"])
```
