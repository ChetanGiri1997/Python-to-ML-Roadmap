# Day 14 — Advanced Functions
 
**Week 3: Functions & Functional Programming**
 
Topics covered: closures, nested functions, higher-order functions, function annotations/type hints, `*args`/`**kwargs` advanced patterns, partial application, recursion with memoization.

### Q1. Nested Function Basics
Write an outer function `outer()` that defines an inner function `inner()` printing a message, and have `outer()` call `inner()` internally. Call `outer()` and observe the output.

### Q2. Introduction to Closures
Write a function `make_counter()` that returns a function `increment()` which remembers and increases a count each time it's called (using a closure over an enclosing variable, not a global). Create two independent counters and demonstrate they don't interfere with each other.

### Q3. Closure with Parameters
Write a function `make_power_func(exponent)` that returns a function `raise_to(base)` computing `base ** exponent`. Create `square_func = make_power_func(2)` and `cube_func = make_power_func(3)` and test both.

### Q4. Modifying Enclosing Scope with `nonlocal`
Write a function `make_accumulator()` that returns an `add(amount)` function which keeps a running total using `nonlocal`. Call `add()` multiple times and print the running total after each call.

### Q5. Higher-Order Functions — Passing Functions as Arguments
Write a function `apply_twice(func, value)` that applies `func` to `value` two times in a row. Test it with a function that doubles a number.
 

### Q6. Higher-Order Functions — Returning Functions
Write a function `get_operation(op_name)` that returns a different function (`add`, `subtract`, `multiply`, or `divide`) depending on the string passed in. Test all four cases.

### Q7. Custom `map` Implementation
Without using Python's built-in `map()`, write your own function `my_map(func, iterable)` that applies `func` to every item in `iterable` and returns a list of results.

### Q8. Custom `filter` Implementation
Without using Python's built-in `filter()`, write your own function `my_filter(func, iterable)` that returns only the items in `iterable` for which `func` returns `True`.

### Q9. Function Annotations / Type Hints
Write a function `add_numbers(a: int, b: int) -> int` with proper type hints. Print its `__annotations__` attribute to show the hints are stored.

### Q10. Default Arguments Referencing Other Parameters (Workaround)
Python doesn't allow default arguments to reference other parameters directly. Write a function `rectangle_area(width, height=None)` where, if `height` is not given, it defaults to being equal to `width` (i.e., a square). Implement this using `None` as a sentinel inside the function body.

### Q11. Recursive Function with Memoization (Manual Dictionary)
Write a recursive `fib_memo(n, cache={})` function that computes Fibonacci numbers efficiently by storing previously computed results in a dictionary. Test it with `n = 30` and compare conceptually (in a comment) to the non-memoized version's speed.

### Q12. Using `functools.lru_cache`
Rewrite a recursive Fibonacci function using the `@lru_cache` decorator from `functools`. Test it with `n = 35`.

### Q13. Partial Function Application
Use `functools.partial` to create a new function `square` from Python's built-in `pow(base, exponent)` function, fixing `exponent=2`. Test `square(5)` and `square(9)`.

### Q14. Variable Scope Chain (LEGB Rule)
Write nested functions (`level1`, `level2`, `level3`) where each level defines a variable with the same name `x`. Print `x` at each level to demonstrate Python's Local-Enclosing-Global-Built-in (LEGB) lookup order.
 

### Q15. Function that Returns Multiple Functions
Write a function `make_calculator()` that returns a dictionary of functions: `{"add": ..., "subtract": ..., "multiply": ..., "divide": ...}`, each being a small inline function. Test calling each one from the dictionary.

### Q16. Recursive Function — Sum of Digits
Write a recursive function `sum_of_digits(n)` that returns the sum of the digits of a non-negative integer `n`. Test with `n = 12345`.
 

### Q17. Recursive Function — Flatten Nested List
Write a recursive function `flatten(nested_list)` that takes a list which may contain other lists (arbitrarily nested) and returns a single flat list of all the elements.

### Q18. Mutual Recursion
Write two functions `is_even(n)` and `is_odd(n)` that call each other recursively (instead of using the modulo operator) to determine if a number is even or odd. Handle the base case for `n == 0`.

### Q19. Higher-Order Function — Function Pipeline
Write a function `pipeline(*functions)` that takes any number of functions and returns a new function which applies them in sequence to an initial value. Example: `pipeline(double, increment)(5)` should first double 5, then increment the result.

### Q20. Decorator Preview — Timing a Function Manually
Without using the `@decorator` syntax yet, write a function `time_it(func, *args)` that records the time before and after calling `func(*args)` using the `time` module, prints how long it took, and returns the function's result.

### Q21. Closures for Data Hiding (Simple "Private" State)
Write a function `make_bank_account(balance)` that returns two functions, `deposit(amount)` and `withdraw(amount)`, both of which modify a shared `balance` variable via closure. Withdraw should prevent the balance from going negative. Demonstrate the balance can only be changed through these two functions.

### Q22. Recursive Binary Search
Write a recursive function `binary_search(sorted_list, target, low=0, high=None)` that returns the index of `target` in `sorted_list`, or `-1` if not found. Test it on a sorted list of at least 10 elements.

### Q23. Currying by Hand
Write a function `curried_add(a)` that returns a function taking `b`, which itself returns a function taking `c`, ultimately returning `a + b + c`. Test it as `curried_add(1)(2)(3)`.

### Q24. Recursive Function — Power Set
Write a recursive function `power_set(s)` that returns all subsets of a given list `s` (the power set), returned as a list of lists.

### Q25. Challenge — Function Factory for Validators
Write a function `make_validator(min_val, max_val)` that returns a function `validate(x)` which returns `True` if `min_val <= x <= max_val`, otherwise `False`. Create at least two different validators with different ranges and test each against several values, including edge cases at the boundaries.
