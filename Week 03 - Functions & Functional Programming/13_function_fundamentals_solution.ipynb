# Day 13 — Function Fundamentals
 
**Week 3: Functions & Functional Programming**
 
Topics covered: defining functions, parameters, return values, default arguments, keyword arguments, `*args`/`**kwargs`, scope (local/global), docstrings.
 

### Q1. Basic Function Definition
Write a function `greet(name)` that returns the string `"Hello, {name}!"`. Call it with three different names and print the results.


### Q2. Function with Default Parameter
Write a function `power(base, exponent=2)` that returns `base` raised to `exponent`. Test it by calling it once with only `base` and once with both arguments.

### Q3. Multiple Return Values
Write a function `min_max(numbers)` that takes a list of numbers and returns both the minimum and maximum value as a tuple. Unpack the result into two variables when calling it.

### Q4. Keyword Arguments
Write a function `build_profile(name, age, city="Unknown", job="Unknown")` that returns a dictionary built from the arguments. Call it using a mix of positional and keyword arguments.

### Q5. Variable-Length Arguments (`*args`)
Write a function `total_sum(*args)` that returns the sum of any number of numeric arguments passed to it. Test it with 0, 1, and 5 arguments.

### Q6. Variable-Length Keyword Arguments (`**kwargs`)
Write a function `print_info(**kwargs)` that prints each key-value pair passed to it in the format `key: value`. Call it with at least 4 keyword arguments.

### Q7. Combining args, *args, and **kwargs
Write a function `order_summary(customer, *items, **details)` that:
- Prints the customer name
- Prints each item ordered
- Prints each extra detail (like `discount=10`, `priority=True`)

### Q8. Local vs Global Scope
Create a global variable `counter = 0`. Write a function `increment()` that increments `counter` by 1 using the `global` keyword. Call it 5 times and print the final value.

### Q9. Scope Without `global`
Write a function that creates a local variable with the same name as a global variable, modifies it inside the function, and prints both the local and global values to demonstrate they are different.

### Q10. Returning Nothing
Write a function `log_message(msg)` that prints a formatted log message but does not explicitly return anything. Print the result of calling this function and explain (in a comment) what Python returns by default.

### Q11. Docstrings
Write a function `celsius_to_fahrenheit(c)` with a proper docstring explaining what it does, its parameter, and its return value. Use `help()` or `.__doc__` to display the docstring.

### Q12. Function as a Value
Write a function `square(x)` that returns `x * x`. Assign this function to a new variable `s` (without calling it) and call `s(5)`. Explain in a comment why this works.

### Q13. Recursive Function — Factorial
Write a recursive function `factorial(n)` that computes the factorial of `n`. Include a base case. Test it with `n = 0, 5, 10`.

### Q14. Recursive Function — Fibonacci
Write a recursive function `fibonacci(n)` that returns the n-th Fibonacci number. Test it for `n = 0` through `n = 10`.

### Q15. Default Mutable Argument Trap
Write a function `add_item(item, lst=[])` that appends `item` to `lst` and returns `lst`. Call it three times with just an item (no list argument) and observe/print the result. In a comment, explain why this happens and how to fix it.

### Q16. Argument Unpacking
Create a list `nums = [3, 7, 2]` and a function `add_three(a, b, c)` that returns their sum. Call the function by unpacking the list using `*nums`.

### Q17. Dictionary Unpacking into Function Calls
Create a dictionary `person = {"name": "Asha", "age": 30}` and a function `display(name, age)`. Call the function by unpacking the dictionary using `**person`.

### Q18. Function Returning a Function (Preview of Closures)
Write a function `make_multiplier(n)` that returns a new function which multiplies its input by `n`. Create `double = make_multiplier(2)` and `triple = make_multiplier(3)`, then test both.

### Q19. Validating Input Inside a Function
Write a function `safe_divide(a, b)` that returns the division result, but if `b == 0`, returns the string `"Cannot divide by zero"` instead of raising an error.

### Q20. Counting Function Calls
Write a function `tracked_function()` that keeps track of how many times it has been called, using a function attribute (e.g., `tracked_function.calls`). Print the call count after calling it 4 times.

### Q21. Real-World Mini Task — Grade Calculator
Write a function `calculate_grade(score)` that returns `"A"` for 90+, `"B"` for 80–89, `"C"` for 70–79, and `"F"` otherwise. Test it with at least 5 different scores.
 

### Q22. Real-World Mini Task — Temperature Converter Suite
Write three functions: `c_to_f(c)`, `f_to_c(f)`, and `c_to_k(c)`. Then write a fourth function `convert(value, from_unit, to_unit)` that uses the appropriate function based on the units given (`"C"`, `"F"`, `"K"`).

### Q23. Function Composition
Write two functions `add_five(x)` and `multiply_by_two(x)`. Write a third function `compose(f, g, x)` that returns `f(g(x))`. Use it to compute `add_five(multiply_by_two(3))` via `compose`.

### Q24. Variadic Average Calculator
Write a function `average(*nums)` that returns the average of any number of arguments. Handle the edge case where no arguments are passed (return `0` instead of raising a `ZeroDivisionError`).

### Q25. Challenge — Flexible Formatter
Write a function `format_record(name, *tags, **fields)` that returns a single string combining:
- The name
- All tags joined by commas
- All fields formatted as `key=value`, joined by spaces
Example call: `format_record("Server1", "prod", "active", region="us-east", cpu=4)` should produce something like:
`"Server1 [prod, active] region=us-east cpu=4"`
