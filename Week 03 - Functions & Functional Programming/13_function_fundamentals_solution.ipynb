### Q6. Variable-Length Keyword Arguments (`**kwargs`)

### Q7. Combining args, *args, and **kwargs

- Prints the customer name
- Prints each item ordered
- Prints each extra detail (like `discount=10`, `priority=True`)

```python
# Define the function
def order_summary(customer, *items, **details):
    print("Customer:", customer)

    print("\nItems Ordered:")
    for item in items:
        print("-", item)

    print("\nDetails:")
    for key, value in details.items():
        print(f"{key}: {value}")

# Call the function with a mix of args, *args, and **kwargs
order_summary(
    "Chetan",
    "Pizza", "Burger", "Pasta",
    discount=10,
    priority=True,
    payment="Card"
)
```

### Q13. Recursive Function — Factorial
Write a recursive function `factorial(n)` that computes the factorial of `n`. Include a base case. Test it with `n = 0, 5, 10`.

```python
# Recursive function to calculate factorial
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    # Recursive case
    return n * factorial(n - 1)

# Test cases
print(factorial(0))
print(factorial(5))
print(factorial(10))
```

### Q14. Recursive Function — Fibonacci
Write a recursive function `fibonacci(n)` that returns the n-th Fibonacci number. Test it for `n = 0` through `n = 10`.

```python
# Recursive Fibonacci function
def fibonacci(n):
    # Base cases
    if n == 0:
        return 0
    if n == 1:
        return 1

    # Recursive case
    return fibonacci(n - 1) + fibonacci(n - 2)

# Test for n = 0 to 10
for i in range(11):
    print(f"fibonacci({i}) =", fibonacci(i))
```

### Q15. Default Mutable Argument Trap
Write a function `add_item(item, lst=[])` that appends `item` to `lst` and returns `lst`. Call it three times with just an item (no list argument) and observe/print the result. In a comment, explain why this happens and how to fix it.

```python
# Function with a mutable default argument
def add_item(item, lst=[]):
    lst.append(item)
    return lst

# Calling the function three times (without passing a list)
print(add_item("A"))
print(add_item("B"))
print(add_item("C"))

# Why this happens:
# The default list [] is created only once when the function is defined.
# So all calls without an explicit list share the SAME list object,
# causing items to accumulate across calls.

# Correct fix (recommended pattern):
# Use None as default and create a new list inside the function.

def add_item_safe(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item_safe("A"))
print(add_item_safe("B"))
print(add_item_safe("C"))
```

### Q18. Function Returning a Function (Preview of Closures)
Write a function `make_multiplier(n)` that returns a new function which multiplies its input by `n`. Create `double = make_multiplier(2)` and `triple = make_multiplier(3)`, then test both.

```python
# Function that returns another function (closure)
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

# Create new functions
double = make_multiplier(2)
triple = make_multiplier(3)

# Test them
print(double(5))   # 5 * 2
print(triple(5))   # 5 * 3
```

### Q22. Real-World Mini Task — Temperature Converter Suite
Write three functions: `c_to_f(c)`, `f_to_c(f)`, and `c_to_k(c)`. Then write a fourth function `convert(value, from_unit, to_unit)` that uses the appropriate function based on the units given (`"C"`, `"F"`, `"K"`).

```python
# Celsius to Fahrenheit
def c_to_f(c):
    return (c * 9/5) + 32

# Fahrenheit to Celsius
def f_to_c(f):
    return (f - 32) * 5/9

# Celsius to Kelvin
def c_to_k(c):
    return c + 273.15


# General conversion function
def convert(value, from_unit, to_unit):
    from_unit = from_unit.upper()
    to_unit = to_unit.upper()

    if from_unit == to_unit:
        return value

    # Convert input to Celsius first (intermediate step)
    if from_unit == "C":
        c = value
    elif from_unit == "F":
        c = f_to_c(value)
    elif from_unit == "K":
        c = value - 273.15
    else:
        return "Invalid from_unit"

    # Convert Celsius to target unit
    if to_unit == "C":
        return c
    elif to_unit == "F":
        return c_to_f(c)
    elif to_unit == "K":
        return c_to_k(c)
    else:
        return "Invalid to_unit"


# Tests
print(convert(0, "C", "F"))
print(convert(100, "C", "K"))
print(convert(32, "F", "C"))
print(convert(300, "K", "F"))
```

### Q23. Function Composition
Write two functions `add_five(x)` and `multiply_by_two(x)`. Write a third function `compose(f, g, x)` that returns `f(g(x))`. Use it to compute `add_five(multiply_by_two(3))` via `compose`.

```python
# Function 1
def add_five(x):
    return x + 5

# Function 2
def multiply_by_two(x):
    return x * 2

# Function composition
def compose(f, g, x):
    return f(g(x))

# Using compose: add_five(multiply_by_two(3))
result = compose(add_five, multiply_by_two, 3)

print(result)
```

### Q24. Variadic Average Calculator
Write a function `average(*nums)` that returns the average of any number of arguments. Handle the edge case where no arguments are passed (return `0` instead of raising a `ZeroDivisionError`).

```python
# Function to compute average of variable number of arguments
def average(*nums):
    if len(nums) == 0:
        return 0
    return sum(nums) / len(nums)

# Test cases
print(average(10, 20, 30))
print(average(5))
print(average())
```

### Q25. Challenge — Flexible Formatter
Write a function `format_record(name, *tags, **fields)` that returns a single string combining:
- The name
- All tags joined by commas
- All fields formatted as `key=value`, joined by spaces
Example call: `format_record("Server1", "prod", "active", region="us-east", cpu=4)` should produce something like:
`"Server1 [prod, active] region=us-east cpu=4"`

```python
# Flexible formatter function
def format_record(name, *tags, **fields):
    # Join tags with commas
    tags_part = f"[{', '.join(tags)}]" if tags else "[]"

    # Format key=value pairs
    fields_part = " ".join(f"{k}={v}" for k, v in fields.items())

    # Combine all parts
    if fields_part:
        return f"{name} {tags_part} {fields_part}"
    else:
        return f"{name} {tags_part}"


# Test case
result = format_record(
    "Server1",
    "prod", "active",
    region="us-east",
    cpu=4
)

print(result)
```
