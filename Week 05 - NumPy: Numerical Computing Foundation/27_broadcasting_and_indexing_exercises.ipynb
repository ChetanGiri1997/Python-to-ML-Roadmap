### Q25. Challenge — Vectorized k-Nearest Neighbors Distance Matrix
Given two sets of points: `A` of shape `(50, 2)` and `B` of shape `(30, 2)`, compute the full Euclidean distance matrix `D` of shape `(50, 30)` where `D[i, j]` is the distance from `A[i]` to `B[j]` — **entirely using broadcasting, no loops**. Then:

- Find all pairs with distance less than a threshold of 0.5
- Find the point in A with the largest minimum distance to any point in B (the most isolated point)
This is the core computation behind kNN classification.
