### Q1. Memory Layout — C vs Fortran Order
Create a 4×4 array in default C order (row-major) and another in Fortran order (`order='F'`). Print the `strides` of each. Explain why a row traversal is faster in C order and a column traversal is faster in Fortran order. Time both traversal patterns on a large array.

### Q2. Understanding Strides
Create a 1D array of integers. Print its `strides` and `itemsize`. Then create a 2D reshape of it and print strides again. Manually predict what the strides should be before printing, then verify.

### Q3. `np.lib.stride_tricks.sliding_window_view`

### Q4. `np.vectorize`

Compare its performance to a direct `np.where` solution on a large array using `%timeit` or `time`. Explain the limitations of `np.vectorize`.

### Q5. `np.frompyfunc`

### Q6. Masked Arrays — Introduction
Create a masked array using `np.ma.array([1, 2, -999, 4, -999, 6], mask=[0,0,1,0,1,0])`. Print the array — masked values show as `--`. Compute mean, sum, and min (they should ignore masked values). Unmask one element and verify the change.

### Q7. Masked Arrays — Creating Masks Dynamically

### Q8. Masked Array Operations
Create two masked arrays with different masks. Add them together and observe how NumPy handles mask propagation (a position masked in either array is masked in the result). Use `np.ma.filled(arr, fill_value=0)` to convert back to a regular array.

### Q9. `np.einsum` — Basics
Use `np.einsum` to compute:
- Dot product of two vectors: `'i,i->'`
- Matrix-vector multiplication: `'ij,j->i'`
- Matrix-matrix multiplication: `'ij,jk->ik'`
- Transpose of a matrix: `'ij->ji'`
- Trace of a matrix: `'ii->'`
Verify each against the standard NumPy equivalent operation.

### Q10. `np.einsum` — Intermediate
Use `np.einsum` to compute:
- Outer product of two vectors: `'i,j->ij'`
- Element-wise multiplication of two matrices then sum all: `'ij,ij->'`
- Batch matrix multiplication (3D): `'bij,bjk->bik'` (multiply a batch of matrices)
- Row-wise dot product: `'ij,ij->i'` (dot product of each row with corresponding row)

### Q11. `np.einsum` Performance
Compare `np.einsum('ij,jk->ik', A, B)` to `A @ B` for matrices of size 200×200. Which is faster? Then compare `np.einsum('ij,ij->i', A, B)` to `(A * B).sum(axis=1)`. Use `timeit` for timing. Explain when `einsum` wins over explicit operations.
