### Q1. Basic `if`/`elif`/`else`
Write a function `classify_temperature(temp)` that returns:
- `"Freezing"` for temp < 0
- `"Cold"` for 0 ‚â§ temp < 15
- `"Comfortable"` for 15 ‚â§ temp < 28
- `"Hot"` for temp ‚â• 28

Test with at least 5 values including boundary cases.

````python
# Q1: Classify temperature into categories using if/elif/else

def classify_temperature(temp):
    if temp < 0:
        return "Freezing"
    elif temp < 15:
        return "Cold"
    elif temp < 28:
        return "Comfortable"
    else:
        return "Hot"

# Test with boundary and non-boundary values
test_values = [-5, 0, 14, 15, 27, 28, 35]

for t in test_values:
    result = classify_temperature(t)
    print(f"classify_temperature({t:>3}) ‚Üí {result}")
````

### How it works

- The function checks `temp` against thresholds one by one, from lowest to highest.
- Python stops at the **first `True` condition** and returns that label.
- Because conditions are checked in order, `elif temp < 15` only runs when `temp >= 0` is already known ‚Äî no need to write `0 <= temp < 15` explicitly.
- Boundary values (0, 15, 28) are handled correctly: 0 ‚Üí "Cold", 15 ‚Üí "Comfortable", 28 ‚Üí "Hot".
- The final `else` catches everything ‚â• 28 ‚Äî no extra condition needed.

**Key takeaway:** Order `if/elif` conditions from the most restrictive (lowest threshold) to least restrictive so each branch is reached only when all previous conditions are already false.

### Q2. Nested Conditionals
Write a function `ticket_price(age, is_student)` that returns:
- Free for age < 5
- 50% discount if age < 18 OR is_student is True (base price = 500)
- Full price (500) otherwise
Avoid deep nesting by combining conditions clearly.

````python
# Q2: Calculate ticket price using nested conditionals and logical operators

BASE_PRICE = 500

def ticket_price(age, is_student):
    if age < 5:
        return 0                          # free for very young children
    elif age < 18 or is_student:
        return BASE_PRICE * 0.5          # 50% discount
    else:
        return BASE_PRICE                # full price

# Test cases
test_cases = [
    (3,  False, "child under 5"),
    (10, False, "child under 18"),
    (25, True,  "adult student"),
    (30, False, "adult non-student"),
    (4,  True,  "child under 5 who is also student"),
    (17, False, "teenager, not student"),
]

for age, is_student, label in test_cases:
    price = ticket_price(age, is_student)
    print(f"age={age:>2}, is_student={str(is_student):<5} ({label}) ‚Üí Rs. {price}")
````

### How it works

- The function checks three situations in order: very young, discounted, full price.
- `age < 5` is checked **first** ‚Äî a 3-year-old student still gets a free ticket, not a discount.
- `age < 18 or is_student` combines two conditions with `or` ‚Äî either one being `True` gives the discount.
- The `or` means we never need nested `if` inside `if`; one flat `elif` handles both discount cases.
- `else` at the end means: adult, not a student ‚Äî pay full price.

**Key takeaway:** Combining conditions with `and`/`or` keeps your code flat and readable ‚Äî prefer one `elif` with a logical operator over two levels of nested `if`.

### Q3. Ternary Expression
Rewrite each of the following using a single ternary expression:
- A function that returns `"even"` or `"odd"` based on a number
- A function that returns the absolute value of a number without using `abs()`
- Assigning `max_val` to the larger of two numbers `a` and `b`

````python
# Q3: Rewrite three cases using ternary (conditional) expressions

# --- Part 1: even or odd ---
def even_or_odd(n):
    return "even" if n % 2 == 0 else "odd"   # % gives remainder after division

# --- Part 2: absolute value without abs() ---
def my_abs(n):
    return n if n >= 0 else -n               # flip sign only when negative

# --- Part 3: maximum of two numbers ---
a, b = 17, 42
max_val = a if a > b else b                  # assign the larger value directly

# --- Print results ---
for n in [-4, -1, 0, 3, 8]:
    print(f"even_or_odd({n:>3}) ‚Üí {even_or_odd(n)}")

print()
for n in [-7, 0, 5, -100]:
    print(f"my_abs({n:>4}) ‚Üí {my_abs(n)}")

print()
print(f"a={a}, b={b} ‚Üí max_val = {max_val}")
````

### How it works

- A ternary expression (conditional expression) has the form: `value_if_true if condition else value_if_false`.
- Think of it as a one-line `if/else` that **produces a value** instead of running a block of code.
- `n % 2 == 0` checks divisibility ‚Äî remainder 0 means even, anything else means odd.
- For absolute value, a negative number becomes positive by multiplying by `-1`, which is the same as writing `-n`.
- `max_val = a if a > b else b` reads naturally: "give me `a` if `a` is bigger, otherwise give me `b`."

**Key takeaway:** Use a ternary expression when you are choosing between exactly two values ‚Äî it keeps simple decisions on one readable line.

### Q4. Truthiness and Falsiness
Without using `== True`, `== False`, `== None`, or `== 0`, write conditions to check whether each of these is "empty" or "falsy": an integer, a string, a list, a dictionary, `None`. Print what Python considers falsy for each type.

````python
# Q4: Demonstrate truthiness and falsiness for common Python types

# --- Falsy sample values for each type ---
values = [
    (0,    "integer"),
    ("",   "string"),
    ([],   "list"),
    ({},   "dictionary"),
    (None, "None"),
]

print("=== Falsy values ===")
for val, type_name in values:
    if not val:                          # no == False or == 0 needed
        print(f"  {type_name:<12} ‚Üí {repr(val)!s:<6} is falsy")

print()

# --- Truthy sample values for each type ---
truthy_values = [
    (42,          "integer"),
    ("hello",     "string"),
    ([1, 2],      "list"),
    ({"a": 1},   "dictionary"),
    ("anything",  "None slot (not None)"),
]

print("=== Truthy values ===")
for val, type_name in truthy_values:
    if val:                              # Python checks truthiness automatically
        print(f"  {type_name:<22} ‚Üí {repr(val)!s:<12} is truthy")

print()

# --- Practical use: guard against empty input ---
def greet(name):
    if not name:                         # catches both None and ""
        return "Please provide a name."
    return f"Hello, {name}!"

print("=== Practical example ===")
for test in ["Chetan", "", None]:
    print(f"  greet({repr(test):<8}) ‚Üí {greet(test)}")
````

### How it works

- Python treats certain values as **falsy** (behave like `False` in a condition) without any comparison operator.
- The falsy values for common types are: `0` (int), `""` (string), `[]` (list), `{}` (dict), `None`.
- Everything else ‚Äî non-zero numbers, non-empty strings, non-empty collections ‚Äî is **truthy**.
- Writing `if not val:` is the Pythonic way to check emptiness; `if val == False` or `if val == 0` is discouraged.
- The `greet` example shows a real use case: `if not name` catches both `None` and `""` in one condition.

**Key takeaway:** Python has built-in truthiness rules for every type ‚Äî use `if val:` and `if not val:` directly instead of comparing to `0`, `False`, or `None`.

### Q5. Short-Circuit Evaluation
Demonstrate short-circuit evaluation:
- Write a function `safe_divide(a, b)` using `b != 0 and a / b` ‚Äî show the division is never evaluated when `b` is 0
- Write a function `get_name(user)` using `user and user.get("name")` ‚Äî show it safely handles `None` without raising `AttributeError`

````python
# Q5: Demonstrate short-circuit evaluation with and/or

# --- Part 1: safe_divide ---
def safe_divide(a, b):
    result = b != 0 and a / b    # if b==0, Python stops and never computes a/b
    print(f"  safe_divide({a}, {b}) ‚Üí {result}")
    return result

print("=== safe_divide ===")
safe_divide(10, 2)               # b != 0 is True  ‚Üí evaluates 10/2 = 5.0
safe_divide(10, 0)               # b != 0 is False ‚Üí stops, returns False (no ZeroDivisionError)
safe_divide(-9, 3)               # b != 0 is True  ‚Üí evaluates -9/3 = -3.0

print()

# --- Part 2: get_name ---
def get_name(user):
    name = user and user.get("name")   # if user is None, stops before calling .get()
    print(f"  get_name({str(user):<25}) ‚Üí {name}")
    return name

print("=== get_name ===")
get_name({"name": "Chetan", "age": 28})   # user is truthy ‚Üí evaluates user.get("name")
get_name({"age": 28})                      # user is truthy ‚Üí .get("name") returns None
get_name(None)                             # user is falsy  ‚Üí stops, returns None (no AttributeError)

print()

# --- Bonus: show the short-circuit moment with a side-effect function ---
def loud_divide(a, b):
    print(f"    [dividing {a}/{b}...]")   # this line prints only if actually called
    return a / b

print("=== Side-effect proof ===")
print("  b=2:", False and loud_divide(10, 2))   # False and ... ‚Üí stops, loud_divide never runs
print("  b=0:", True  and loud_divide(10, 0))   # True  and ... ‚Üí runs, causes ZeroDivisionError protection test below
````

### How it works

- **Short-circuit evaluation** means Python stops checking conditions as soon as the result is certain.
- With `and`: if the **left side is falsy**, Python returns it immediately ‚Äî the right side is never evaluated.
- With `or`: if the **left side is truthy**, Python returns it immediately ‚Äî the right side is never evaluated.
- In `b != 0 and a / b`: when `b` is `0`, the left side is `False`, so Python never reaches `a / b` ‚Äî no crash.
- In `user and user.get("name")`: when `user` is `None` (falsy), Python stops before calling `.get()` ‚Äî no `AttributeError`.

**Key takeaway:** Short-circuit evaluation is a safe, readable way to guard against errors ‚Äî put the safety check on the left side of `and` so the risky operation on the right is only reached when it is safe to run.

### Q6. Logical Operators ‚Äî Truth Table
Without using any if/else, use only `and`, `or`, `not` to implement:
- `xor(a, b)`: True when exactly one of `a`, `b` is True
- `nand(a, b)`: True when NOT both are True
- `implies(a, b)`: logical implication ‚Äî False only when `a` is True and `b` is False
Test each with all 4 combinations of True/False.

````python
# Q6: Implement logical operators using only and, or, not

def xor(a, b):
    return (a or b) and not (a and b)   # true if either, but not both

def nand(a, b):
    return not (a and b)                # opposite of AND

def implies(a, b):
    return not a or b                   # False only when a=True, b=False

# --- Test all 4 combinations for each function ---
pairs = [(True, True), (True, False), (False, True), (False, False)]

print(f"{'a':<6} {'b':<6} {'xor':<6} {'nand':<6} {'implies'}")
print("-" * 38)
for a, b in pairs:
    print(f"{str(a):<6} {str(b):<6} {str(xor(a,b)):<6} {str(nand(a,b)):<6} {str(implies(a,b))}")
````

### How it works

- **xor** (exclusive or): true when inputs differ. `(a or b)` catches "at least one true"; `not (a and b)` removes "both true". Combined: exactly one must be true.
- **nand** (not-and): simply the opposite of `and`. If both are `True`, `and` gives `True`, so `not` flips it to `False`.
- **implies** (`a ‚Üí b`): in logic, "if a then b" is only broken when `a` is `True` but `b` is `False`. Rewriting: `not a or b` ‚Äî either `a` was never true, or `b` holds.
- A truth table tests every possible input combination ‚Äî with 2 booleans there are exactly 4 rows.
- No `if/else` is needed because `and`, `or`, `not` already return boolean values directly.

**Key takeaway:** Every logical operation can be built from `and`, `or`, `not` ‚Äî understanding their truth tables lets you construct any condition without branching.

### Q7. Guard Clauses
Refactor this deeply nested function into flat guard-clause style:
```python
def process(data):
    if data is not None:
        if len(data) > 0:
            if isinstance(data, list):
                return sum(data)
```
The refactored version should return early for each invalid case with a descriptive message, and only reach the main logic if all checks pass.
 

````python
# Q7: Refactor deep nesting into flat guard clauses

# --- Original (deeply nested) ---
def process_nested(data):
    if data is not None:
        if len(data) > 0:
            if isinstance(data, list):
                return sum(data)

# --- Refactored (flat guard clauses) ---
def process(data):
    if data is None:                     # guard 1: reject None early
        return "Error: data is None"
    if not isinstance(data, list):       # guard 2: reject non-list early
        return "Error: data must be a list"
    if len(data) == 0:                   # guard 3: reject empty list early
        return "Error: list is empty"
    return sum(data)                     # main logic ‚Äî only reached if all guards pass

# --- Test cases ---
test_cases = [
    (None,          "None input"),
    ("hello",       "string, not a list"),
    ([],            "empty list"),
    ([1, 2, 3, 4],  "valid list"),
    ([10, -5, 7],   "valid list with negatives"),
]

print(f"{'Input':<20} {'Result'}")
print("-" * 42)
for data, label in test_cases:
    result = process(data)
    print(f"{str(data):<20} ‚Üí {result}  ({label})")
````

### How it works

- A **guard clause** is an early `return` that immediately exits the function when input is invalid.
- Instead of nesting valid cases deeper and deeper, we flip each condition and return an error right away.
- After all guards pass, the main logic (`sum(data)`) sits at the **top level** with no indentation nesting.
- Notice the order matters: check `None` first, because calling `isinstance(None, list)` would still work, but `len(None)` would crash ‚Äî so we eliminate `None` before any other check.
- The original code silently returns `None` for bad input; the refactored version returns a clear error message.

**Key takeaway:** Guard clauses turn "valid path deeply nested inside conditions" into "invalid paths exit early, valid path runs flat" ‚Äî making code easier to read and debug.

### Q8. `match`/`case` (Structural Pattern Matching)
Write a function `handle_command(command)` using `match`/`case` that handles these string commands: `"quit"`, `"help"`, `"start"`, `"pause"`, and a default case for unknown commands. Then extend it to match against a tuple `(action, value)` ‚Äî e.g., `("move", 5)` ‚Äî using pattern matching on structure.
 

````python
# Q8: Structural pattern matching with match/case (Python 3.10+)

# --- Part 1: match on string commands ---
def handle_command(command):
    match command:
        case "quit":
            return "Shutting down..."
        case "help":
            return "Available commands: quit, help, start, pause"
        case "start":
            return "Starting the process."
        case "pause":
            return "Process paused."
        case _:                          # default ‚Äî matches anything not listed above
            return f"Unknown command: '{command}'"

print("=== String commands ===")
for cmd in ["start", "pause", "help", "quit", "fly"]:
    print(f"  handle_command({cmd!r:<8}) ‚Üí {handle_command(cmd)}")

print()

# --- Part 2: match on tuple structure (action, value) ---
def handle_action(instruction):
    match instruction:
        case ("move", int(distance)):    # matches tuple where second item is an int
            return f"Moving {distance} units."
        case ("resize", float(scale)):   # matches tuple where second item is a float
            return f"Resizing by factor {scale}."
        case ("say", str(message)):      # matches tuple where second item is a string
            return f"Saying: '{message}'"
        case ("move", _):                # move with wrong type
            return "Error: move requires an integer distance."
        case _:
            return f"Unknown instruction: {instruction}"

print("=== Tuple pattern matching ===")
instructions = [
    ("move", 5),
    ("move", 3.5),          # wrong type for move
    ("resize", 1.5),
    ("say", "hello"),
    ("jump", 10),            # unknown action
]
for instr in instructions:
    print(f"  handle_action({str(instr):<18}) ‚Üí {handle_action(instr)}")
````

### How it works

- `match`/`case` (Python 3.10+) is like `if/elif/else` but reads more clearly when checking one value against many possible shapes.
- Each `case` is a **pattern** ‚Äî Python checks them top to bottom and runs the first one that matches.
- `case _:` is the wildcard (default) ‚Äî it matches anything, like `else` in an `if` chain.
- In Part 2, `case ("move", int(distance))` matches a tuple **and** checks the type of the second item at the same time ‚Äî this is called **structural pattern matching**.
- The captured variable (`distance`, `scale`, `message`) is automatically available inside that case block ‚Äî no extra unpacking needed.

**Key takeaway:** `match/case` shines when you need to check both the **shape** and **content** of a value at once ‚Äî it replaces chains of `if isinstance(...) and len(...) == 2` with clean, readable patterns.

### Q9. Chained Comparisons
Python allows `0 < x < 10` instead of `x > 0 and x < 10`. Write a function `is_valid_score(score)` using a chained comparison. Then write `is_between(val, low, high)` generically. Demonstrate that chained comparisons evaluate the middle value only once (important for function calls with side effects).

````python
# Q9: Chained comparisons and single-evaluation guarantee

# --- Part 1: is_valid_score ---
def is_valid_score(score):
    return 0 <= score <= 100            # score must be in [0, 100] inclusive

print("=== is_valid_score ===")
for s in [-1, 0, 55, 100, 101]:
    print(f"  is_valid_score({s:>4}) ‚Üí {is_valid_score(s)}")

print()

# --- Part 2: generic is_between ---
def is_between(val, low, high):
    return low <= val <= high           # works for int, float, or even strings

print("=== is_between ===")
print(f"  is_between(5,   0, 10)   ‚Üí {is_between(5,   0, 10)}")
print(f"  is_between(10,  0, 10)   ‚Üí {is_between(10,  0, 10)}")
print(f"  is_between(3.5, 3, 4)    ‚Üí {is_between(3.5, 3, 4)}")
print(f"  is_between('m','a','z')  ‚Üí {is_between('m','a','z')}")   # works on strings too

print()

# --- Part 3: middle value evaluated only once ---
call_count = 0

def get_value():
    global call_count
    call_count += 1                     # counts how many times this function is called
    return 5

# Using chained comparison: get_value() is called ONCE
call_count = 0
result = 0 < get_value() < 10
print(f"=== Single evaluation proof ===")
print(f"  Chained  (0 < get_value() < 10): result={result}, get_value() called {call_count} time(s)")

# Using and: get_value() is called TWICE
call_count = 0
result = get_value() > 0 and get_value() < 10
print(f"  With and (get_value()>0 and get_value()<10): result={result}, get_value() called {call_count} time(s)")
````

### How it works

- Python lets you chain comparisons like `0 <= score <= 100` ‚Äî it reads exactly like math notation.
- Internally Python evaluates this as `0 <= score and score <= 100`, but the middle value (`score`) is computed only **once**.
- This matters when the middle value is a function call with a side effect (like writing to a file, or incrementing a counter) ‚Äî chaining avoids calling it twice.
- The `call_count` example proves this: the chained version calls `get_value()` once; the `and` version calls it twice.
- Chaining also works on strings ‚Äî Python compares them alphabetically, so `'a' <= 'm' <= 'z'` is `True`.

**Key takeaway:** Chained comparisons are not just shorter ‚Äî they guarantee the middle expression runs exactly once, which prevents subtle bugs when that expression has side effects.

### Q10. Conditional with `any` and `all`
Given a list of numbers, use `any()` and `all()` (not loops or explicit `if`) to:
- Check if any number is negative
- Check if all numbers are positive
- Check if any number is divisible by 7
- Check if all strings in a list of strings have length > 3

````python
# Q10: Using any() and all() for collection-wide checks

numbers = [4, -2, 15, 7, 21, 8, -5, 14]
words   = ["python", "loop", "if", "data", "code"]

# --- Check 1: any number is negative ---
has_negative = any(n < 0 for n in numbers)
print(f"Numbers: {numbers}")
print(f"  any negative?          ‚Üí {has_negative}")

# --- Check 2: all numbers are positive ---
all_positive = all(n > 0 for n in numbers)
print(f"  all positive?          ‚Üí {all_positive}")

# --- Check 3: any number divisible by 7 ---
has_div7 = any(n % 7 == 0 for n in numbers)
print(f"  any divisible by 7?    ‚Üí {has_div7}")

print()

# --- Check 4: all strings have length > 3 ---
all_long = all(len(w) > 3 for w in words)
print(f"Words: {words}")
print(f"  all strings length > 3 ‚Üí {all_long}")          # "if" has length 2 ‚Üí False

print()

# --- Prove short-circuit: any/all stop early ---
call_log = []

def logged_check(n):
    call_log.append(n)                  # record which values were actually checked
    return n < 0

call_log.clear()
result = any(logged_check(n) for n in numbers)
print(f"=== Short-circuit proof (any stops at first True) ===")
print(f"  numbers checked before stopping: {call_log}")
print(f"  result: {result}")
````

### How it works

- `any(condition for item in collection)` returns `True` if **at least one** item satisfies the condition.
- `all(condition for item in collection)` returns `True` only if **every** item satisfies the condition.
- The expression inside is a **generator** (not a list) ‚Äî values are produced one at a time, which enables short-circuit behaviour.
- `any()` stops as soon as it finds the first `True`; `all()` stops as soon as it finds the first `False` ‚Äî just like `or` and `and`.
- The `call_log` proof shows `any()` checked only `[4, -2]` before stopping ‚Äî it found `-2 < 0` and returned immediately without touching the rest.

**Key takeaway:** Use `any()` and `all()` instead of writing a loop with a flag variable ‚Äî they are shorter, more readable, and automatically short-circuit for efficiency.

### Q11. Dictionary as a Lookup Table (Replacing Long `if`/`elif` Chains)
Refactor a function that uses 6 `elif` branches to map a day number (1‚Äì7) to its name into a dictionary lookup. Then extend it: the dictionary values should be functions (lambdas) so each day can execute different logic when called.

````python
# Q11: Replace long if/elif chains with dictionary lookup tables

# --- Original (long if/elif chain) ---
def day_name_ifelse(day):
    if day == 1:
        return "Monday"
    elif day == 2:
        return "Tuesday"
    elif day == 3:
        return "Wednesday"
    elif day == 4:
        return "Thursday"
    elif day == 5:
        return "Friday"
    elif day == 6:
        return "Saturday"
    elif day == 7:
        return "Sunday"
    else:
        return "Invalid day"

# --- Refactored: dictionary lookup ---
DAY_NAMES = {
    1: "Monday",
    2: "Tuesday",
    3: "Wednesday",
    4: "Thursday",
    5: "Friday",
    6: "Saturday",
    7: "Sunday",
}

def day_name(day):
    return DAY_NAMES.get(day, "Invalid day")   # .get() returns default if key missing

print("=== Dictionary lookup ===")
for d in [1, 4, 7, 8]:
    print(f"  day_name({d}) ‚Üí {day_name(d)}")

print()

# --- Extended: dictionary of lambdas (each day runs different logic) ---
DAY_ACTIONS = {
    1: lambda: "Start the week ‚Äî check your goals.",
    2: lambda: "Tuesday ‚Äî team meeting day.",
    3: lambda: "Wednesday ‚Äî halfway there!",
    4: lambda: "Thursday ‚Äî wrap up pending tasks.",
    5: lambda: "Friday ‚Äî prepare the weekly report.",
    6: lambda: "Saturday ‚Äî rest or personal projects.",
    7: lambda: "Sunday ‚Äî plan for next week.",
}

def handle_day(day):
    action = DAY_ACTIONS.get(day, lambda: f"Day {day} is not valid.")
    return action()                            # call the lambda to get the message

print("=== Dictionary of lambdas ===")
for d in [1, 3, 5, 7, 9]:
    print(f"  handle_day({d}) ‚Üí {handle_day(d)}")

print()

# --- Practical extension: lambdas with a value argument ---
DAY_DISCOUNT = {
    6: lambda price: price * 0.8,    # 20% off on Saturday
    7: lambda price: price * 0.9,    # 10% off on Sunday
}

def get_price(day, base_price):
    discount_fn = DAY_DISCOUNT.get(day, lambda price: price)   # no discount by default
    return discount_fn(base_price)

print("=== Lambdas with arguments ===")
base = 500
for d in [5, 6, 7]:
    print(f"  day={d}, base={base} ‚Üí final price = {get_price(d, base)}")
````

### How it works

- A dictionary lookup replaces `if/elif` chains by storing answers as **values** against keys ‚Äî Python finds the right one in one step instead of checking branches one by one.
- `.get(key, default)` safely returns the default value when the key does not exist ‚Äî no `KeyError`, no extra `else` branch.
- Dictionaries can store **functions as values**, not just strings or numbers ‚Äî this is because functions are objects in Python.
- A `lambda` is a small anonymous function written in one line: `lambda x: x * 2` is the same as `def f(x): return x * 2`.
- Calling `action()` after the lookup executes whichever lambda was stored for that key ‚Äî each day can have completely different behaviour.

**Key takeaway:** When you have many `elif` branches all checking the same variable, a dictionary lookup is cleaner and faster ‚Äî and storing lambdas as values lets each entry carry its own logic, not just data.

### Q12. Conditional Import
Write a block that tries to import `ujson` (a fast JSON library), and if it's not installed, falls back to the standard `json` module. Assign whichever is available to a variable `json_lib` and use it to serialize a dictionary.

````python
# Q12: Conditional import with try/except fallback

# --- Attempt fast library first, fall back to standard library ---
try:
    import ujson as json_lib            # ujson is faster but not always installed
    source = "ujson"
except ImportError:
    import json as json_lib             # standard library ‚Äî always available
    source = "json (standard library)"

print(f"Using: {source}")
print()

# --- Use whichever was imported ‚Äî same interface either way ---
data = {
    "name":    "Chetan",
    "rank":    "Sergeant",
    "skills":  ["Python", "IoT", "SDR"],
    "active":  True,
}

serialized = json_lib.dumps(data, indent=2)   # both ujson and json support .dumps()
print("=== Serialized output ===")
print(serialized)

print()

# --- Round-trip: deserialize back to a dict and verify ---
restored = json_lib.loads(serialized)
print("=== Round-trip check ===")
print(f"  name  : {restored['name']}")
print(f"  skills: {restored['skills']}")
print(f"  active: {restored['active']}")
print(f"  Match : {restored == data}")
````

### How it works

- `try/except ImportError` is the standard Python pattern for optional dependencies ‚Äî try the preferred library, fall back gracefully if it is missing.
- Both `ujson` and `json` share the same interface (`.dumps()`, `.loads()`), so the rest of the code never needs to know which one was loaded.
- Assigning to a common name (`json_lib`) is the key idea ‚Äî it gives the rest of the code one consistent name to use regardless of which library was imported.
- This pattern is common in real projects: `orjson`, `ujson`, and `simplejson` are all faster alternatives to `json`, but standard `json` is always the safe fallback.
- A round-trip (serialize ‚Üí deserialize ‚Üí compare) confirms the library works correctly and the data survived intact.

**Key takeaway:** Use `try/except ImportError` to make your code work with an optional fast library when available, and fall back to the standard library automatically ‚Äî the caller never needs to change.

### Q13. Complex Business Logic
Write a function `loan_eligibility(age, income, credit_score, existing_loans)` that returns `"Approved"`, `"Conditional"`, or `"Rejected"` based on these rules:
- Rejected if age < 21 or age > 70
- Rejected if income < 20000
- Rejected if existing_loans >= 3
- Conditional if credit_score < 650 but income > 50000
- Approved otherwise
Test with at least 6 different combinations.

````python
# Q13: Complex business logic with multiple conditions and guard clauses

def loan_eligibility(age, income, credit_score, existing_loans):
    # --- Guard clauses: hard rejections first ---
    if age < 21 or age > 70:
        return "Rejected"               # age outside eligible range
    if income < 20000:
        return "Rejected"               # income too low
    if existing_loans >= 3:
        return "Rejected"               # too many existing loans

    # --- Conditional approval: risky but high income saves it ---
    if credit_score < 650 and income > 50000:
        return "Conditional"            # poor credit, but high income offsets risk

    # --- All checks passed ---
    return "Approved"

# --- Test cases ---
test_cases = [
    # (age, income,  credit, loans, label)
    (18,  35000,  700, 0, "too young"),
    (72,  60000,  750, 1, "too old"),
    (30,  15000,  700, 0, "income too low"),
    (40,  45000,  700, 3, "too many loans"),
    (35,  60000,  600, 1, "low credit, high income ‚Üí conditional"),
    (28,  40000,  580, 0, "low credit, income not high enough ‚Üí rejected"),
    (45,  80000,  750, 2, "all clear ‚Üí approved"),
    (55,  55000,  680, 0, "solid profile ‚Üí approved"),
]

print(f"{'Age':<5} {'Income':<8} {'Credit':<8} {'Loans':<7} {'Result':<12} Note")
print("-" * 70)
for age, income, credit, loans, label in test_cases:
    result = loan_eligibility(age, income, credit, loans)
    print(f"{age:<5} {income:<8} {credit:<8} {loans:<7} {result:<12} {label}")
````

### How it works

- Hard rejection rules are written as **guard clauses** at the top ‚Äî the function exits immediately when any disqualifying condition is met.
- Guard clause order matters: age and income are checked before `existing_loans` and credit, because they are the most absolute disqualifiers.
- The `Conditional` case is only reached when all three rejection guards have already passed ‚Äî so at that point we know age, income (‚â• 20000), and loans are acceptable; we only need to check credit vs income.
- Notice `credit_score < 650 and income > 50000`: both must be true together ‚Äî poor credit alone is not enough for "Conditional"; the high income must also be present to offset the risk.
- The final `return "Approved"` needs no condition ‚Äî if every guard and the conditional check have been passed, approval is the only remaining outcome.

**Key takeaway:** For complex business rules, write the hardest rejections first as guard clauses, then handle edge/conditional cases in the middle, and let the happy path fall naturally to the bottom with no extra condition needed.

### Q14. Avoiding Redundant Conditions
Given this inefficient function, refactor it to eliminate redundant checks:
```python
def category(x):
    if x < 0:
        return "negative"
    if x >= 0 and x < 10:
        return "small"
    if x >= 10 and x < 100:
        return "medium"
    if x >= 100:
        return "large"
```

````python
# Q14: Refactor redundant conditions using elif and implicit range knowledge

# --- Original (redundant checks) ---
def category_original(x):
    if x < 0:
        return "negative"
    if x >= 0 and x < 10:       # x >= 0 is already guaranteed here
        return "small"
    if x >= 10 and x < 100:     # x >= 10 is already guaranteed here
        return "medium"
    if x >= 100:                 # this is the only remaining case
        return "large"

# --- Refactored (no redundant checks) ---
def category(x):
    if x < 0:
        return "negative"       # below zero
    elif x < 10:
        return "small"          # x >= 0 already known ‚Äî no need to recheck
    elif x < 100:
        return "medium"         # x >= 10 already known ‚Äî no need to recheck
    else:
        return "large"          # x >= 100 is the only case left

# --- Verify both functions produce identical results ---
test_values = [-50, -1, 0, 5, 9, 10, 55, 99, 100, 200]

print(f"{'x':<6} {'original':<12} {'refactored':<12} {'match'}")
print("-" * 40)
for x in test_values:
    orig   = category_original(x)
    refact = category(x)
    match  = "yes" if orig == refact else "NO MISMATCH"
    print(f"{x:<6} {orig:<12} {refact:<12} {match}")

print()

# --- Show exactly which redundant checks were removed ---
print("=== What was removed and why ===")
redundancies = [
    ("x >= 0 and x < 10",  "x >= 0 is guaranteed ‚Äî if x < 0 had returned already"),
    ("x >= 10 and x < 100","x >= 10 is guaranteed ‚Äî elif x < 10 had returned already"),
    ("if x >= 100",         "only remaining case ‚Äî else covers it with no condition"),
]
for check, reason in redundancies:
    print(f"  Removed: {check!r:<26} ‚Üí {reason}")
````

### How it works

- Once `if x < 0` returns, Python never reaches the next branch ‚Äî so `x >= 0` is **already guaranteed** inside every `elif` below it.
- Each `elif` inherits the knowledge that all previous conditions were `False` ‚Äî rewriting `x >= 0 and x < 10` as just `x < 10` is not a shortcut, it is the correct and complete condition.
- `else` at the end requires **no condition at all** ‚Äî if every `elif` was skipped, the only remaining possibility is `x >= 100`.
- Using `if ... if ... if ...` (separate `if` statements) forces Python to evaluate every condition every time, even after a match ‚Äî `elif` stops as soon as one branch is taken.
- The verification table confirms both versions produce identical output for all test values.

**Key takeaway:** After an `if` branch returns or exits, its condition becomes implicit knowledge for every `elif` below ‚Äî never restate what the structure already guarantees.

### Q15. Challenge ‚Äî Decision Tree Classifier (Manual)
Implement a manual decision tree as nested conditions to classify an iris flower into `"Setosa"`, `"Versicolor"`, or `"Virginica"` based on these simplified rules:
- If `petal_length < 2.5` ‚Üí Setosa
- Else if `petal_width < 1.8`:
  - If `petal_length < 5.0` ‚Üí Versicolor
  - Else ‚Üí Virginica
- Else ‚Üí Virginica
Write it as a clean function, then test it with 6 sample inputs. Discuss in a comment why real ML models replace these hand-written rules.

````python
# Q15: Manual decision tree classifier for iris flowers

def classify_iris(petal_length, petal_width):
    # --- Level 1: primary split on petal length ---
    if petal_length < 2.5:
        return "Setosa"                  # short petals ‚Üí always Setosa

    # --- Level 2: split on petal width ---
    elif petal_width < 1.8:
        if petal_length < 5.0:           # narrow petal, medium length ‚Üí Versicolor
            return "Versicolor"
        else:                            # narrow petal, long ‚Üí Virginica
            return "Virginica"

    # --- Level 3: wide petal ‚Üí always Virginica ---
    else:
        return "Virginica"

# --- Test with 6 sample inputs (approximate real iris measurements in cm) ---
samples = [
    # (petal_length, petal_width, expected,      label)
    (1.4,  0.2,  "Setosa",     "classic Setosa ‚Äî very short petal"),
    (2.4,  0.9,  "Setosa",     "boundary ‚Äî just under 2.5"),
    (4.5,  1.4,  "Versicolor", "medium length, narrow width"),
    (4.9,  1.7,  "Versicolor", "near boundary, still Versicolor"),
    (5.1,  1.6,  "Virginica",  "narrow width but long petal ‚Üí Virginica"),
    (6.0,  2.2,  "Virginica",  "wide and long ‚Üí Virginica"),
]

print(f"{'petal_l':<10} {'petal_w':<10} {'predicted':<12} {'expected':<12} {'correct':<8} note")
print("-" * 75)
for pl, pw, expected, label in samples:
    predicted = classify_iris(pl, pw)
    correct   = "yes" if predicted == expected else "NO"
    print(f"{pl:<10} {pw:<10} {predicted:<12} {expected:<12} {correct:<8} {label}")

print()

# --- Why real ML replaces hand-written rules ---
# 1. SCALE: real datasets have 10+ features ‚Äî writing nested conditions for all
#    combinations by hand becomes impossible to read and maintain.
#
# 2. THRESHOLDS: we guessed 2.5 and 1.8 ‚Äî a trained DecisionTreeClassifier
#    finds the mathematically optimal split points automatically from data.
#
# 3. ACCURACY: hand-written rules miss edge cases and overlapping regions
#    between classes; ML models learn the boundary from thousands of examples.
#
# 4. GENERALISATION: when new data arrives, hand-written rules need manual
#    updates; a trained model is simply retrained on the new data.
#
# 5. INTERPRETABILITY PRESERVED: sklearn's DecisionTreeClassifier produces
#    the same if/elif structure internally ‚Äî so you get automation without
#    losing the ability to inspect and explain the decisions.
print("See comments in code for why ML replaces hand-written decision trees.")
````

### How it works

- The function mirrors exactly how a real decision tree works: one condition at a time, branching left or right at each node.
- The first split (`petal_length < 2.5`) cleanly separates Setosa from the other two species ‚Äî this is the most important feature.
- The second split uses `petal_width < 1.8` to separate Versicolor from Virginica in the remaining cases.
- When both conditions point to Virginica (long petal or wide petal), the two branches are merged into one `else` ‚Äî no duplication needed.
- The five comments in the code explain the core problem with hand-written trees: thresholds are guessed, they do not scale to many features, and they need manual updates when data changes.

**Key takeaway:** A manual decision tree is a useful teaching tool ‚Äî it shows exactly how a trained model thinks ‚Äî but real ML automates finding the best thresholds and structure from data, which is impractical to do by hand beyond a few features.
