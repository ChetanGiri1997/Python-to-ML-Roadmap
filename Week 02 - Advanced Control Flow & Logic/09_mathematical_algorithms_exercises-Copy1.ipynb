# Day 09 — Mathematical Algorithms
 
**Week 2: Advanced Control Flow & Logic**
 
Topics covered: number theory (primes, GCD, LCM, factorials), iterative and recursive algorithms, modular arithmetic, digit manipulation, sequence generation, mathematical series, `math` module.
 
---

### Q1. Prime Checker
Write a function `is_prime(n)` that returns `True` if `n` is prime. Optimize it to only check divisors up to `√n` using `math.isqrt(n)`. Test it for `n = 0, 1, 2, 13, 97, 100, 999983`.

### Q2. Sieve of Eratosthenes
Implement the Sieve of Eratosthenes to find all primes up to `n`. Return them as a list. Test with `n = 50` and `n = 200`. Compare the output count to the known prime-counting values.

### Q3. GCD and LCM
Implement `gcd(a, b)` using the Euclidean algorithm (iterative, no `math.gcd`). Then implement `lcm(a, b)` using the relationship `lcm = a * b // gcd(a, b)`. Extend both to accept any number of arguments using `functools.reduce`. Verify against `math.gcd`.

### Q4. Factorial — Iterative and Recursive
Implement `factorial_iterative(n)` using a loop and `factorial_recursive(n)` using recursion. Both should handle `n = 0`. Compare their call behavior for large `n` — at what `n` does the recursive version hit Python's recursion limit?

### Q5. Fibonacci Sequence — Three Ways
Implement Fibonacci three ways:
- Iterative (loop)
- Recursive (naive)
- Using a generator (`yield`)
Generate the first 15 Fibonacci numbers with each approach and verify they match.

### Q6. Digit Manipulation
Write functions to:
- `sum_of_digits(n)`: return the sum of digits of a positive integer (e.g., `1234` → `10`)
- `reverse_digits(n)`: reverse the digits of an integer (e.g., `1234` → `4321`)
- `is_palindrome_number(n)`: check if an integer reads the same forwards and backwards
- `count_digits(n)`: count the number of digits without using `len(str(n))`


### Q7. Armstrong Numbers
A number is an Armstrong number (narcissistic number) if the sum of its digits each raised to the power of the number of digits equals the number itself (e.g., `153 = 1³ + 5³ + 3³`). Write a function `is_armstrong(n)` and find all Armstrong numbers between 1 and 9999.

### Q8. Modular Arithmetic
Demonstrate modular arithmetic:
- `fast_power(base, exp, mod)`: compute `base^exp % mod` efficiently using Python's built-in 3-argument `pow()`, then implement it manually using repeated squaring
- Write a function `is_divisible_by_9(n)` that uses the digit-sum rule (sum of digits divisible by 9) rather than `n % 9`
- Find all numbers from 1 to 100 that are divisible by 3 or 7 but not both


### Q9. Number Base Conversion
Write functions (without using built-in `bin()`, `oct()`, `hex()`):
- `to_binary(n)`: convert a positive integer to its binary string
- `to_base(n, base)`: convert to any base from 2 to 16 (using `"0123456789ABCDEF"`)
- `from_base(s, base)`: convert a string in the given base back to decimal
Verify: `from_base(to_base(255, 16), 16) == 255`.

### Q10. Series Summation
Compute the following series to `n` terms (write a general function for each):
- Sum of natural numbers: `1 + 2 + 3 + ... + n` (then verify with the formula `n*(n+1)/2`)
- Sum of squares: `1² + 2² + ... + n²`
- Sum of cubes: `1³ + 2³ + ... + n³`
- Alternating series: `1 - 1/2 + 1/3 - 1/4 + ...` (approximates `ln(2)` — compare for large `n`)

### Q11. Perfect Numbers and Divisors
A perfect number equals the sum of its proper divisors (e.g., 6 = 1+2+3). Write:
- `get_divisors(n)`: return all proper divisors of `n`
- `is_perfect(n)`: check if `n` is perfect
- Find all perfect numbers below 10,000


### Q12. Collatz Conjecture
The Collatz sequence starting from `n`: if `n` is even, divide by 2; if odd, multiply by 3 and add 1; repeat until reaching 1. Write a function `collatz_length(n)` returning the number of steps to reach 1. Find the starting number below 1000 that produces the longest Collatz sequence.
 

### Q13. Pascal's Triangle
Write a function `pascals_triangle(n)` that generates the first `n` rows of Pascal's Triangle as a list of lists. Print it in a visually centered format. Use it to extract: the 5th row, the diagonal of 1s, and the triangular numbers (second diagonal).

### Q14. Mathematical Series — π and e Approximation
Approximate mathematical constants:
- π using the Leibniz formula: `π/4 = 1 - 1/3 + 1/5 - 1/7 + ...`
- e using the series: `e = 1 + 1/1! + 1/2! + 1/3! + ...`
For each, show the approximation after 10, 100, 1000, and 10000 terms. Compare to `math.pi` and `math.e`.

### Q15. Challenge — Prime Factorization and Cryptography Basics
Write a function `prime_factors(n)` that returns the prime factorization of `n` as a dictionary `{prime: exponent}` (e.g., `360 → {2:3, 3:2, 5:1}`). Then:
- Verify by reconstructing `n` from the factors
- Use it to compute `lcm` of a list of numbers via their factorizations
- Implement a basic RSA key-generation demo: pick two small primes `p` and `q`, compute `n = p*q` and `φ(n) = (p-1)*(q-1)`, find a valid public exponent `e` (coprime with `φ(n)`), compute the private key `d` using `pow(e, -1, φ(n))`. Encrypt and decrypt the integer message `42`.
