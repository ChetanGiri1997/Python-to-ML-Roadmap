### Q1. String Methods — Cleaning
Given the messy string `"  Hello,   World!  This is   Python.  "`:
- Strip leading and trailing whitespace
- Collapse multiple internal spaces into one (use `split()` then `join()`)
- Convert to title case
- Count how many words it contains after cleaning

```python
# Q1: Clean a messy string — strip, collapse spaces, title case, word count

messy = "  Hello,   World!  This is   Python.  "

# Step 1: Strip leading and trailing whitespace
stripped = messy.strip()

# Step 2: Collapse multiple internal spaces — split() splits on any whitespace
#         join() puts the words back together with a single space
cleaned = " ".join(stripped.split())

# Step 3: Convert to title case
titled = cleaned.title()

# Step 4: Count words
word_count = len(titled.split())

print("Original :", repr(messy))
print("Cleaned  :", repr(titled))
print("Word count:", word_count)
```

### How it works

- `.strip()` removes spaces (and newlines) from both ends of the string.
- `.split()` with no argument splits on **any whitespace** and throws away empty parts — so multiple spaces become one split.
- `" ".join(...)` puts the words back together with exactly one space between each.
- `.title()` capitalises the first letter of every word.
- `.split()` on the cleaned string gives a list of words; `len()` counts them.

**Key takeaway:** `split()` + `join()` is the standard Python trick for collapsing any amount of internal whitespace into a single space.


### Q2. `split` and `join`
Given a CSV-like string `"Alice,28,Kathmandu,Engineer"`:
- Split into a list on commas
- Access each field by index and print with a label
- Rejoin with ` | ` as the separator
- Split a multiline string (use `\n`) into individual lines and strip each line


```python
# Q2: split and join — CSV parsing, field access, rejoin, multiline split

csv_string = "Alice,28,Kathmandu,Engineer"

# Step 1: Split on commas
fields = csv_string.split(",")

# Step 2: Access each field by index and print with a label
print("Name    :", fields[0])
print("Age     :", fields[1])
print("City    :", fields[2])
print("Job     :", fields[3])

# Step 3: Rejoin with | as separator
pipe_string = "|".join(fields)
print("\nRejoined:", pipe_string)

# Step 4: Split a multiline string into lines and strip each line
multiline = "  Alice  \n  28  \n  Kathmandu  \n  Engineer  "
lines = multiline.split("\n")                   # split on newline character
stripped_lines = [line.strip() for line in lines]  # strip each line individually

print("\nMultiline lines (stripped):")
for line in stripped_lines:
    print(" -", line)
```

### How it works

- `.split(",")` cuts the string at every comma and returns a list of parts.
- Each part is accessed by its index — `fields[0]` is the first item, `fields[3]` is the last.
- `"|".join(fields)` is the reverse of split — it glues the list back into one string using `|` as the glue.
- `.split("\n")` cuts a multiline string at every newline, giving one element per line.
- `.strip()` on each line removes any extra spaces around the text.

**Key takeaway:** `split` and `join` are opposites — `split` breaks a string into a list, `join` builds a string from a list.


### Q3. `find`, `index`, `count`, `in`
Using the string `"the quick brown fox jumps over the lazy dog"`:
- Find the position of the first occurrence of `"the"` using `find`
- Find the position of the second `"the"` (hint: use `find` with a `start` argument)
- Count how many times `"the"` appears (case-insensitive)
- Use `in` to check if `"fox"` and `"cat"` are present

```python
# Q3: find, index, count, in — searching inside strings

sentence = "the quick brown fox jumps over the lazy dog"

# Step 1: Find position of the FIRST "the"
first_pos = sentence.find("the")
print("First 'the' at index:", first_pos)

# Step 2: Find position of the SECOND "the"
# pass first_pos + 1 as the start argument to search after the first match
second_pos = sentence.find("the", first_pos + 1)
print("Second 'the' at index:", second_pos)

# Step 3: Count occurrences — case-insensitive
# .lower() makes sure "The", "THE", "the" all count the same
count = sentence.lower().count("the")
print("Total 'the' count (case-insensitive):", count)

# Step 4: Check presence with 'in'
print("\n'fox' in sentence:", "fox" in sentence)
print("'cat' in sentence:", "cat" in sentence)
```

### How it works

- `.find("the")` scans left to right and returns the index of the first match. It returns `-1` if not found (unlike `.index()` which raises an error).
- `.find("the", start)` begins scanning from position `start`, so passing `first_pos + 1` skips past the first match to find the second.
- `.lower().count("the")` first converts the whole string to lowercase, then counts every non-overlapping match — making the search case-insensitive.
- The `in` operator returns `True` or `False` and is the simplest way to check if a substring exists anywhere in a string.

**Key takeaway:** Use `find` when you need the position of a match, `count` when you need how many times it appears, and `in` when you only need a yes/no answer.


### Q4. `startswith`, `endswith`, and `replace`
Given a list of filenames `["report.pdf", "data.csv", "notes.txt", "image.PNG", "script.py", "README.md"]`:
- Filter only `.csv` and `.txt` files using `endswith`
- Filter filenames that start with a capital letter using `startswith` with a tuple
- Rename all `.PNG` extensions to `.png` using `replace`

```python
# Q4: startswith, endswith, replace — filtering and renaming filenames

filenames = ["report.pdf", "data.csv", "notes.txt", "image.PNG", "script.py", "README.md"]

# Step 1: Filter .csv and .txt files
# endswith() accepts a tuple — checks if the string ends with ANY of the given suffixes
data_files = [f for f in filenames if f.endswith((".csv", ".txt"))]
print("CSV and TXT files:", data_files)

# Step 2: Filter filenames that start with a capital letter
# pass a tuple of all capital letters to startswith()
capitals = tuple("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
capital_files = [f for f in filenames if f.startswith(capitals)]
print("Start with capital:", capital_files)

# Step 3: Rename .PNG to .png using replace
# replace(old, new) swaps every occurrence of old with new
renamed = [f.replace(".PNG", ".png") for f in filenames]
print("After renaming PNG:", renamed)
```

### How it works

- `.endswith((".csv", ".txt"))` accepts a **tuple of suffixes** and returns `True` if the string ends with any one of them — no need to write two separate checks.
- `.startswith(capitals)` works the same way with a tuple — here we build the tuple from all 26 uppercase letters using `tuple("ABC...")`.
- `.replace(".PNG", ".png")` scans the string and swaps every exact match of `.PNG` with `.png`. Since only `"image.PNG"` contains `.PNG`, only that filename changes.
- List comprehensions (`[f for f in filenames if ...]`) give a clean one-line way to filter a list.

**Key takeaway:** Both `startswith` and `endswith` accept a tuple of options, so you can check multiple prefixes or suffixes in a single call.

### Q5. String Slicing
Given `s = "Python Programming"`:
- Extract `"Python"` using a slice
- Extract `"Programming"` using a slice
- Reverse the string using a slice
- Extract every other character
- Extract the string backwards, skipping every other character

```python
# Q5: String slicing — extract, reverse, step

s = "Python Programming"

# Step 1: Extract "Python" — characters from index 0 up to (not including) 6
python_part = s[:6]
print("'Python'      :", python_part)

# Step 2: Extract "Programming" — starts at index 7 to the end
programming_part = s[7:]
print("'Programming' :", programming_part)

# Step 3: Reverse the string — step of -1 walks backwards through every character
reversed_s = s[::-1]
print("Reversed      :", reversed_s)

# Step 4: Every other character — step of 2 skips one character each time
every_other = s[::2]
print("Every other   :", every_other)

# Step 5: Backwards, skipping every other character — step of -2
backwards_skip = s[::-2]
print("Backwards skip:", backwards_skip)
```

### How it works

- A slice has the form `s[start:stop:step]`. Any part can be omitted — omitting `start` means "from the beginning", omitting `stop` means "to the end".
- `s[:6]` gives characters at indices 0, 1, 2, 3, 4, 5 — the `stop` index (6) is **not included**.
- `s[7:]` starts at index 7 (the `P` in `Programming`) and goes to the end.
- `s[::-1]` uses a step of `-1` — it starts from the last character and walks backwards one step at a time, reversing the string.
- `s[::2]` uses a step of `2` — it picks every second character going forward.
- `s[::-2]` combines both ideas — walks backwards and skips every other character.

**Key takeaway:** The `step` value in a slice controls direction and how many characters to jump — negative step reverses, larger step skips.

### Q6. f-Strings — Formatting Numbers and Alignment
Using f-strings:
- Format the number `3.14159265` to exactly 3 decimal places
- Format `1234567.89` with comma thousands separator and 2 decimal places
- Left-align, center, and right-align the string `"hello"` in a field of width 20 (using different fill characters)
- Format an integer in binary, octal, and hexadecimal

```python
# Q6: f-strings — number formatting, alignment, base conversion

# Step 1: Float to exactly 3 decimal places
# :.3f means "fixed-point number with 3 digits after the decimal"
pi = 3.14159265
print(f"Pi (3 dp)        : {pi:.3f}")

# Step 2: Thousands separator + 2 decimal places
# :,.2f — comma as thousands separator, fixed 2 decimal places
big_number = 1234567.89
print(f"Big number       : {big_number:,.2f}")

# Step 3: Alignment in a field of width 20
word = "hello"
# < = left-align, fill remaining space with -
print(f"Left  (-fill)    : |{word:-<20}|")
# ^ = center, fill with *
print(f"Center (*fill)   : |{word:*^20}|")
# > = right-align, fill with .
print(f"Right  (.fill)   : |{word:.>20}|")

# Step 4: Integer in binary, octal, and hexadecimal
n = 255
print(f"\nDecimal          : {n}")
print(f"Binary  (base 2) : {n:b}")
print(f"Octal   (base 8) : {n:o}")
print(f"Hex     (base 16): {n:x}")   # lowercase hex
print(f"Hex     (base 16): {n:X}")   # uppercase hex
```

### How it works

- Inside an f-string, `{value:format_spec}` controls how the value is displayed. The part after `:` is the **format spec**.
- `:.3f` — `f` means fixed-point float; `.3` sets 3 digits after the decimal point.
- `:,.2f` — the `,` adds a thousands separator before the rest of the float format.
- For alignment, the format spec is `fill_char direction width` — e.g. `-<20` means fill with `-`, align left, total width 20. `^` centres, `>` right-aligns.
- `:b`, `:o`, `:x` convert an integer to binary, octal, and hexadecimal respectively. No separate function call needed.

**Key takeaway:** The f-string format spec `{value:fill direction width .precision type}` is a mini language — learn the order once and you can format almost anything inline.

### Q7. `ord` and `chr` — Character Encoding
- Print the ASCII/Unicode code points for `'A'`, `'a'`, `'0'`, and `' '` using `ord()`
- Use `chr()` to print the alphabet (a–z) in a single line
- Implement a `caesar_cipher(text, shift)` function that shifts each letter by `shift` positions, wrapping around, and leaves non-letter characters unchanged
- Encode and then decode a message to verify round-trip correctness

```python
# Q7: ord and chr — character encoding and Caesar cipher

# Step 1: Print code points for key characters
print("Code points:")
print(f"  ord('A') = {ord('A')}")
print(f"  ord('a') = {ord('a')}")
print(f"  ord('0') = {ord('0')}")
print(f"  ord(' ') = {ord(' ')}")

# Step 2: Print a–z in one line using chr()
# ord('a') = 97, ord('z') = 122 — so range(97, 123) covers all lowercase letters
alphabet = " ".join(chr(i) for i in range(ord('a'), ord('z') + 1))
print("\nAlphabet:", alphabet)

# Step 3: Caesar cipher function
def caesar_cipher(text, shift):
    result = []
    for ch in text:
        if ch.isupper():
            # shift within A–Z: subtract ord('A') to get 0–25, apply shift, wrap with % 26
            shifted = chr((ord(ch) - ord('A') + shift) % 26 + ord('A'))
            result.append(shifted)
        elif ch.islower():
            # same logic for a–z
            shifted = chr((ord(ch) - ord('a') + shift) % 26 + ord('a'))
            result.append(shifted)
        else:
            # spaces, digits, punctuation stay unchanged
            result.append(ch)
    return "".join(result)

# Step 4: Encode and decode to verify round-trip
message = "Hello, Kathmandu! 123"
shift   = 7

encoded = caesar_cipher(message, shift)
decoded = caesar_cipher(encoded, -shift)   # reverse shift to decode

print("\nOriginal :", message)
print("Encoded  :", encoded)
print("Decoded  :", decoded)
print("Round-trip OK:", message == decoded)
```

### How it works

- `ord(ch)` returns the integer code point of a character. `chr(n)` does the reverse — it gives the character for that code point.
- The alphabet `a–z` lives at code points 97–122. `range(ord('a'), ord('z') + 1)` steps through all of them.
- The Caesar cipher shifts each letter by `shift` positions. The key formula is:
  - Subtract the base (`ord('A')` or `ord('a')`) to get a number from 0 to 25.
  - Add the shift, then use `% 26` to wrap around past Z back to A.
  - Add the base back to get the final code point, then `chr()` converts it to a character.
- Non-letter characters (spaces, digits, punctuation) are appended unchanged.
- To decode, apply the same function with `-shift` — the opposite direction.

**Key takeaway:** `ord` and `chr` let you treat characters as numbers, which makes arithmetic on letters (like shifting) straightforward.

### Q8. `str.maketrans` and `translate`
- Use `str.maketrans` and `str.translate` to implement ROT-13 encoding (shift every letter by 13)
- Use `translate` to remove all vowels from a string (pass `None` as the replacement)
- Use `translate` to replace multiple characters at once (e.g., replace `'aeiou'` with `'*****'`)

```python
# Q8: str.maketrans and str.translate — ROT-13, remove vowels, replace chars

# Step 1: ROT-13 using maketrans + translate
# Build the mapping: a–z shifted by 13, A–Z shifted by 13
lowercase = "abcdefghijklmnopqrstuvwxyz"
uppercase = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# maketrans(from_chars, to_chars) maps each char in from_chars to the same-position char in to_chars
rot13_table = str.maketrans(
    lowercase + uppercase,
    lowercase[13:] + lowercase[:13] + uppercase[13:] + uppercase[:13]
)

message = "Hello, Kathmandu!"
encoded = message.translate(rot13_table)
decoded = encoded.translate(rot13_table)   # ROT-13 is its own inverse

print("Original :", message)
print("ROT-13   :", encoded)
print("Decoded  :", decoded)

# Step 2: Remove all vowels — pass None as the value to delete those characters
vowels = "aeiouAEIOU"
delete_table = str.maketrans("", "", vowels)   # third argument = chars to delete
no_vowels = "Python Programming".translate(delete_table)
print("\nNo vowels:", no_vowels)

# Step 3: Replace each vowel with '*'
# maketrans(from_chars, to_chars) — lengths must match
replace_table = str.maketrans("aeiouAEIOU", "**********")
starred = "Python Programming".translate(replace_table)
print("Starred  :", starred)
```

### How it works

- `str.maketrans(from_chars, to_chars)` builds a lookup table. Every character in `from_chars` maps to the character at the same position in `to_chars`.
- `.translate(table)` applies that table to every character in the string in one pass — much faster than a loop of `.replace()` calls.
- For ROT-13, the alphabet is split at position 13 and the two halves are swapped: `abcdefghijklm` → `nopqrstuvwxyz` and vice versa. Because the shift is exactly half of 26, encoding and decoding use the same table.
- `str.maketrans("", "", chars_to_delete)` uses the optional **third argument** — characters listed there are deleted (mapped to nothing).
- To replace with `*`, `to_chars` must have the same length as `from_chars`, so `"**********"` (10 stars) matches the 10 vowels.

**Key takeaway:** `maketrans` + `translate` is the fastest way to do many single-character substitutions or deletions in one pass — cleaner than chaining multiple `.replace()` calls.

### Q9. Multiline and Raw Strings
- Create a multiline string using triple quotes containing at least 3 lines of a poem or text
- Demonstrate a raw string `r"..."` by showing that `\n` and `\t` are treated as literal backslash characters
- Show the difference between `"C:\new_folder"` and `r"C:\new_folder"` (Windows path issue)
- Use `textwrap.dedent` to remove common leading whitespace from a multiline string


```python
# Q9: Multiline strings, raw strings, and textwrap.dedent

import textwrap

# Step 1: Multiline string with triple quotes
poem = """Roses are red,
Violets are blue,
Python is awesome,
And so are you."""

print("Multiline string:")
print(poem)

# Step 2: Raw string — backslash sequences are NOT interpreted
normal = "First line\nSecond line\tTabbed"
raw    = r"First line\nSecond line\tTabbed"

print("\nNormal string  :", normal)   # \n and \t are interpreted
print("Raw string     :", raw)        # \n and \t printed as-is

# Step 3: Windows path problem and solution
# \n inside a path is interpreted as a newline — this is a common bug
bad_path  = "C:\new_folder\table"
good_path = r"C:\new_folder\table"

print("\nBad path (\\n and \\t interpreted):")
print(bad_path)
print("Good path (raw string, safe):")
print(good_path)

# Step 4: textwrap.dedent — remove common leading whitespace
# Indented multiline string (common when written inside a function or if-block)
indented = """
    The mountains rise high,
    The rivers run deep,
    Karnali flows free,
    The valleys quietly sleep.
    """

dedented = textwrap.dedent(indented).strip()   # strip() removes the leading/trailing blank lines
print("\nAfter dedent:")
print(dedented)
```

### How it works

- Triple quotes (`"""` or `'''`) let a string span multiple lines without any special characters — line breaks in the source code become `\n` in the string.
- A raw string (`r"..."`) tells Python to ignore all backslash sequences. Every `\` is stored as a literal backslash character, so `\n` is two characters, not a newline.
- Windows file paths use `\`, and letters like `n`, `t`, `r` after a `\` create escape sequences (`\n` = newline, `\t` = tab). Using `r"..."` for paths avoids this bug entirely.
- `textwrap.dedent` looks at all lines in the string, finds the **common leading whitespace**, and removes it from every line — useful when multiline strings are indented inside a function or class.

**Key takeaway:** Use raw strings (`r"..."`) for file paths and regex patterns where backslashes must be literal; use `textwrap.dedent` to clean up indentation in multiline strings written inside indented code blocks.

### Q10. String Immutability
Demonstrate that strings are immutable by:
- Trying `s[0] = "X"` and catching the `TypeError`
- Showing that `s.replace("a", "b")` returns a NEW string and leaves the original unchanged
- Explaining (in a comment) why `result = ""` inside a loop using `+=` is inefficient and what to use instead (`"".join(list)`)

```python
# Q10: String immutability — TypeError, replace returns new string, += vs join

s = "kathmandu"

# Step 1: Try to modify a character in-place — this will always fail
try:
    s[0] = "X"
except TypeError as e:
    print("Cannot modify string in-place:")
    print(f"  TypeError: {e}")

# Step 2: .replace() returns a NEW string — original is unchanged
modified = s.replace("a", "@")

print(f"\nOriginal  : {s}")        # s is still "kathmandu"
print(f"Modified  : {modified}")   # modified is the new string

# confirm they are different objects in memory
print(f"Same object? {s is modified}")

# Step 3: += in a loop vs join

# INEFFICIENT — += creates a brand new string on every iteration
# because strings are immutable, Python cannot extend the old one in-place.
# With N characters, this creates N intermediate strings and copies more data each time.
result = ""
for ch in s:
    result += ch.upper()
print(f"\nUsing +=   : {result}")

# EFFICIENT — collect parts in a list, join once at the end
# list.append() is fast (no copying), and "".join() does a single allocation
parts = []
for ch in s:
    parts.append(ch.upper())
result_fast = "".join(parts)
print(f"Using join : {result_fast}")

# For this short string both are instant — the difference matters for large strings
# (thousands of characters) where += becomes noticeably slow
```

### How it works

- Strings in Python are **immutable** — once created, no character inside them can be changed. `s[0] = "X"` raises a `TypeError` because index assignment is not allowed.
- `.replace()` does not touch the original string. It creates and returns a **new string** with the substitution applied. The original variable `s` still holds the old value.
- `s is modified` checks whether both variables point to the exact same object in memory — `False` confirms they are two separate strings.
- The `+=` inefficiency: because a string cannot grow in-place, `result += ch` must allocate a brand-new string and copy everything across on every loop iteration. For a string of length N, that is roughly N² copy operations total.
- `list.append()` + `"".join()` avoids this — appending to a list is fast, and `join` makes exactly one new string at the end.

**Key takeaway:** Strings are immutable — every "change" creates a new string. When building a string inside a loop, always collect parts in a list and use `"".join()` at the end.

### Q11. Palindrome Checker
Write a function `is_palindrome(s)` that returns `True` if `s` is a palindrome. Make it case-insensitive and ignore spaces and punctuation. Test it with: `"racecar"`, `"A man a plan a canal Panama"`, `"hello"`, `"Was it a car or a cat I saw"`.

```python
# Q11: Palindrome checker — case-insensitive, ignores spaces and punctuation

def is_palindrome(s):
    # Step 1: Lowercase and keep only letters and digits
    # isalnum() returns True for letters and digits, False for spaces/punctuation
    cleaned = "".join(ch.lower() for ch in s if ch.isalnum())

    # Step 2: Compare cleaned string to its reverse
    return cleaned == cleaned[::-1]

# Test cases
test_cases = [
    "racecar",
    "A man a plan a canal Panama",
    "hello",
    "Was it a car or a cat I saw",
]

for text in test_cases:
    result = is_palindrome(text)
    print(f"{result!s:<5}  |  {text}")
```

### How it works

- `ch.isalnum()` returns `True` only for letters and digits — it filters out spaces, commas, and all other punctuation in one step.
- `.lower()` on each character makes the comparison case-insensitive.
- `"".join(...)` assembles the filtered characters into a clean string with no gaps.
- `cleaned[::-1]` reverses the cleaned string using a slice with step `-1`.
- If the cleaned string reads the same forwards and backwards, it is a palindrome.

**Key takeaway:** The palindrome check itself is just one comparison (`s == s[::-1]`) — the real work is cleaning the input first so spaces and capitalisation do not interfere.

### Q12. Basic Regular Expressions with `re`
Import `re` and demonstrate:
- `re.search(pattern, string)` to find if a pattern exists
- `re.findall(pattern, string)` to extract all matches
- `re.sub(pattern, replacement, string)` to replace matches
- `re.split(pattern, string)` to split on a pattern
Use these to: extract all numbers from a sentence, validate an email format (simple pattern), and remove all punctuation from text.

```python
# Q12: Basic regex with re — search, findall, sub, split

import re

# ── re.search ──────────────────────────────────────────────────────────────
# re.search returns a match object if the pattern is found anywhere, else None
text = "Order #4821 was shipped on 2025-06-15."

match = re.search(r"\d{4}-\d{2}-\d{2}", text)   # pattern: date like YYYY-MM-DD
if match:
    print("Date found    :", match.group())       # .group() gives the matched text
    print("At position   :", match.start(), "–", match.end())

# ── re.findall ─────────────────────────────────────────────────────────────
# re.findall returns a list of ALL non-overlapping matches
sentence = "I have 3 cats, 12 fish, and 100 books on 2 shelves."
numbers = re.findall(r"\d+", sentence)           # \d+ matches one or more digits
print("\nAll numbers   :", numbers)

# ── re.sub ─────────────────────────────────────────────────────────────────
# re.sub replaces every match with the replacement string
dirty = "Hello!!!   World???  This... is-- Python!!"
# \s+ matches one or more whitespace characters; replace runs of spaces with one space
cleaned = re.sub(r"\s+", " ", dirty).strip()
print("\nCleaned spaces:", cleaned)

# ── re.split ───────────────────────────────────────────────────────────────
# re.split splits a string at every match of the pattern
csv_like = "Alice;28,Kathmandu|Engineer"
parts = re.split(r"[;,|]", csv_like)            # split on ; or , or |
print("\nSplit parts   :", parts)

# ── Practical task 1: extract all numbers from a sentence ──────────────────
info = "The temperature is -3.5 degrees and altitude is 1400 meters."
all_numbers = re.findall(r"-?\d+\.?\d*", info)  # optional minus, digits, optional decimal
print("\nExtracted numbers:", all_numbers)

# ── Practical task 2: simple email validation ──────────────────────────────
def is_valid_email(email):
    # pattern: word chars → @ → word chars → dot → 2–4 letters
    pattern = r"^\w+@\w+\.\w{2,4}$"
    return bool(re.search(pattern, email))

emails = ["user@example.com", "bad-email@", "name@domain.org", "missing_at_sign"]
for email in emails:
    print(f"  {is_valid_email(email)!s:<5} {email}")

# ── Practical task 3: remove all punctuation ──────────────────────────────
punctuated = "Hello, World! Python's great... isn't it?"
no_punct = re.sub(r"[^\w\s]", "", punctuated)  # keep only word chars and spaces
print("\nNo punctuation:", no_punct)
```

### How it works

- `re.search(pattern, text)` scans the whole string and returns a **match object** for the first match, or `None` if not found. Call `.group()` on the result to get the matched text.
- `re.findall(pattern, text)` returns a **list** of every non-overlapping match — no match object needed.
- `re.sub(pattern, replacement, text)` replaces every match with the replacement string and returns the new string.
- `re.split(pattern, text)` works like `.split()` but the separator is a regex — `[;,|]` means "split on any one of these three characters".
- Key pattern pieces used here: `\d` = digit, `\w` = letter/digit/underscore, `\s` = whitespace, `+` = one or more, `*` = zero or more, `^`/`$` = start/end of string, `[^...]` = any character NOT in the set.

**Key takeaway:** The four core `re` functions cover almost every common need — `search` (exists?), `findall` (extract all), `sub` (replace), `split` (divide); learn their return types and you can handle most text-parsing tasks.

### Q13. String Parsing — Extracting Structured Data
Given a log-format string:
`"[2024-06-15 14:32:01] ERROR user_service.py:87 — Connection timeout after 30s"`
 
Write code to extract (without using regex at first, then optionally with regex):
- The date and time
- The log level
- The filename and line number
- The error message

```python
# Q13: Parse a structured log string — manual methods first, then regex

log = "[2024-06-15 14:32:01] ERROR user_service.py:87 — Connection timeout after 30s"

# ── Method 1: Manual parsing with string methods ───────────────────────────

# Step 1: Extract date and time — they sit between [ and ]
bracket_content = log[log.find("[") + 1 : log.find("]")]   # slice between [ and ]
date, time = bracket_content.split(" ")
print("Date      :", date)
print("Time      :", time)

# Step 2: Extract log level — first word after the closing bracket
after_bracket = log[log.find("]") + 2:]                    # skip "] "
log_level = after_bracket.split(" ")[0]
print("Log level :", log_level)

# Step 3: Extract filename and line number — second word after the bracket
location = after_bracket.split(" ")[1]                     # "user_service.py:87"
filename, line_number = location.split(":")
print("Filename  :", filename)
print("Line no.  :", line_number)

# Step 4: Extract error message — everything after the em dash "—"
error_message = log.split("—")[1].strip()
print("Message   :", error_message)

# ── Method 2: Same extraction using regex ─────────────────────────────────
import re

print("\n--- regex version ---")

pattern = (
    r"\[(\d{4}-\d{2}-\d{2}) "   # capture date:  YYYY-MM-DD
    r"(\d{2}:\d{2}:\d{2})\] "   # capture time:  HH:MM:SS
    r"(\w+) "                    # capture log level: word
    r"([\w.]+):(\d+) "          # capture filename and line number
    r"— (.+)"                    # capture message: everything after —
)

m = re.search(pattern, log)
if m:
    print("Date      :", m.group(1))
    print("Time      :", m.group(2))
    print("Log level :", m.group(3))
    print("Filename  :", m.group(4))
    print("Line no.  :", m.group(5))
    print("Message   :", m.group(6))
```

### How it works

**Manual method:**
- `log.find("[")` and `log.find("]")` locate the bracket positions; slicing between them gives the timestamp.
- `.split(" ")[0]` and `[1]` pick individual space-separated tokens after the bracket.
- `.split(":") ` separates filename from line number since `:` divides them.
- `.split("—")[1].strip()` cuts at the em dash and trims whitespace to isolate the message.

**Regex method:**
- Parentheses `(...)` in a pattern create **capture groups** — `m.group(1)` returns the first group, `m.group(2)` the second, and so on.
- `\d{4}-\d{2}-\d{2}` matches a date; `[\w.]+` matches a filename that may contain dots; `.+` greedily matches everything to the end for the message.
- Regex is more compact but harder to read — the manual approach is easier to debug when the format is simple and well-known.

**Key takeaway:** For structured strings with predictable separators, manual methods (`find`, `split`, slicing) are often clearer; use regex when the pattern is irregular or you need to match multiple formats at once.

### Q14. String Compression — Run-Length Encoding
Implement a `run_length_encode(s)` function that compresses a string by replacing consecutive repeated characters with the character followed by its count. Example: `"aaabbbccddddee"` → `"a3b3c2d4e2"`. Then implement `run_length_decode(s)` to reverse it. Verify round-trip for several strings.


```python
# Q14: Run-length encoding and decoding

def run_length_encode(s):
    if not s:
        return ""

    result = []
    count  = 1

    # Walk through the string starting from the second character
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:        # same as previous — extend the run
            count += 1
        else:
            # run ended — record the character and its count
            result.append(s[i - 1])
            result.append(str(count))
            count = 1               # reset for the next run

    # Append the final run (loop exits before recording the last one)
    result.append(s[-1])
    result.append(str(count))

    return "".join(result)


def run_length_decode(s):
    if not s:
        return ""

    result = []
    i      = 0

    while i < len(s):
        char = s[i]                  # letter comes first
        i += 1

        # Collect all digit characters that follow (handles counts >= 10)
        num_str = ""
        while i < len(s) and s[i].isdigit():
            num_str += s[i]
            i += 1

        count = int(num_str)
        result.append(char * count)  # repeat the character count times

    return "".join(result)


# ── Tests ──────────────────────────────────────────────────────────────────
test_cases = [
    "aaabbbccddddee",
    "aabbcc",
    "abcd",              # no repeats — every count is 1
    "aaaaaaaaaa",        # single character repeated 10 times
    "aabbbaaacc",
]

print(f"{'Original':<20} {'Encoded':<20} {'Decoded':<20} {'OK'}")
print("-" * 65)
for original in test_cases:
    encoded = run_length_encode(original)
    decoded = run_length_decode(encoded)
    ok      = "yes" if decoded == original else "NO"
    print(f"{original:<20} {encoded:<20} {decoded:<20} {ok}")
```

### How it works

**Encoding:**
- Walk the string one character at a time. Compare each character to the previous one.
- If they match, increment a counter. If they differ, record the previous character and its count, then reset the counter to 1.
- The final run must be appended after the loop ends, since the loop exits before it records the last group.

**Decoding:**
- Read one letter, then read all following digit characters (could be more than one digit for counts ≥ 10).
- Convert the digit string to an integer and use `char * count` to repeat the character — `"a" * 3` gives `"aaa"`.
- Move the index forward manually with `i` instead of a `for` loop, because each step consumes a variable number of characters.

**Key takeaway:** Run-length encoding is a simple compression technique — it only helps when the string has long runs of repeated characters; for strings with no repeats, the encoded form is actually twice as long.

### Q15. Challenge — Text Statistics Tool
Write a function `text_stats(text)` that returns a dictionary with:
- `word_count`: total number of words
- `unique_words`: number of unique words (case-insensitive)
- `avg_word_length`: average length of words (rounded to 2 decimal places)
- `longest_word`: the longest word
- `most_common_word`: the most frequently occurring word (case-insensitive, ignore punctuation)
- `sentence_count`: number of sentences (split on `.`, `!`, `?`)
- `char_frequency`: a dictionary of letter frequencies (lowercase, letters only)
Test it on a paragraph of at least 5 sentences.

```python
# Q15: Text statistics tool

import re

def text_stats(text):
    # ── Word list — lowercase, punctuation stripped from each word ──────────
    raw_words   = text.split()
    clean_words = [re.sub(r"[^\w]", "", w).lower() for w in raw_words]
    clean_words = [w for w in clean_words if w]          # drop empty strings

    # ── Word count ──────────────────────────────────────────────────────────
    word_count  = len(clean_words)

    # ── Unique words ────────────────────────────────────────────────────────
    unique_words = len(set(clean_words))

    # ── Average word length ─────────────────────────────────────────────────
    avg_word_length = round(sum(len(w) for w in clean_words) / word_count, 2)

    # ── Longest word ────────────────────────────────────────────────────────
    longest_word = max(clean_words, key=len)

    # ── Most common word — count manually with a dict ───────────────────────
    freq = {}
    for word in clean_words:
        freq[word] = freq.get(word, 0) + 1
    most_common_word = max(freq, key=freq.get)

    # ── Sentence count — split on . ! ? and remove empty parts ──────────────
    sentences    = re.split(r"[.!?]", text)
    sentence_count = len([s for s in sentences if s.strip()])

    # ── Character frequency — letters only, lowercase ───────────────────────
    char_frequency = {}
    for ch in text.lower():
        if ch.isalpha():
            char_frequency[ch] = char_frequency.get(ch, 0) + 1
    # sort by frequency descending for readability
    char_frequency = dict(sorted(char_frequency.items(), key=lambda x: x[1], reverse=True))

    return {
        "word_count"      : word_count,
        "unique_words"    : unique_words,
        "avg_word_length" : avg_word_length,
        "longest_word"    : longest_word,
        "most_common_word": most_common_word,
        "sentence_count"  : sentence_count,
        "char_frequency"  : char_frequency,
    }


# ── Test paragraph ──────────────────────────────────────────────────────────
paragraph = """
Nepal is a beautiful country nestled in the Himalayas.
The mountains of Nepal attract thousands of tourists every year.
Kathmandu is the capital city and the cultural heart of Nepal.
People in Nepal are known for their warmth and hospitality.
The rivers of Nepal are perfect for white-water rafting and adventure sports!
"""

stats = text_stats(paragraph)

# ── Print results ───────────────────────────────────────────────────────────
print(f"Word count       : {stats['word_count']}")
print(f"Unique words     : {stats['unique_words']}")
print(f"Avg word length  : {stats['avg_word_length']}")
print(f"Longest word     : {stats['longest_word']}")
print(f"Most common word : {stats['most_common_word']}")
print(f"Sentence count   : {stats['sentence_count']}")
print(f"\nChar frequency (top 10):")
top_10 = list(stats["char_frequency"].items())[:10]
for ch, count in top_10:
    bar = "█" * count
    print(f"  {ch} : {bar} {count}")
```

### How it works

- `re.sub(r"[^\w]", "", w)` strips punctuation from each word by removing any character that is not a letter, digit, or underscore — so `"Nepal."` becomes `"nepal"` after `.lower()`.
- `set(clean_words)` automatically removes duplicates; its length gives the unique word count.
- `sum(len(w) for w in clean_words) / word_count` computes the average word length — a generator expression avoids building an intermediate list.
- `max(clean_words, key=len)` finds the longest word by comparing lengths directly.
- Word frequencies are counted into a plain dict using `.get(word, 0) + 1`. `max(freq, key=freq.get)` picks the key with the highest value.
- `re.split(r"[.!?]", text)` splits on any sentence-ending punctuation; filtering empty strings handles trailing punctuation correctly.
- `char_frequency` is sorted by count descending using `sorted(..., key=lambda x: x[1], reverse=True)` so the most frequent letters appear first.

**Key takeaway:** A statistics tool like this is built from small, focused steps — clean the data first, then compute each metric independently; combining simple string operations with a plain frequency dict covers most text analysis needs without any special libraries.
