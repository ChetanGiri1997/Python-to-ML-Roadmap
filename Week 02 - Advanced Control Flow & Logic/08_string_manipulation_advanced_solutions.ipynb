### Q11. Palindrome Checker

### Q12. Basic Regular Expressions with `re`
Import `re` and demonstrate:
- `re.search(pattern, string)` to find if a pattern exists

- `re.sub(pattern, replacement, string)` to replace matches
- `re.split(pattern, string)` to split on a pattern
Use these to: extract all numbers from a sentence, validate an email format (simple pattern), and remove all punctuation from text.

### Q14. String Compression — Run-Length Encoding

### Q15. Challenge — Text Statistics Tool
Write a function `text_stats(text)` that returns a dictionary with:
- `word_count`: total number of words
- `unique_words`: number of unique words (case-insensitive)
- `avg_word_length`: average length of words (rounded to 2 decimal places)
- `longest_word`: the longest word
- `most_common_word`: the most frequently occurring word (case-insensitive, ignore punctuation)
- `sentence_count`: number of sentences (split on `.`, `!`, `?`)
- `char_frequency`: a dictionary of letter frequencies (lowercase, letters only)
Test it on a paragraph of at least 5 sentences.
