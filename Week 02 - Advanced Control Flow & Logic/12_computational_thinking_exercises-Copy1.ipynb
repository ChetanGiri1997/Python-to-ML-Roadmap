# Day 12 — Computational Thinking (Week 2 Capstone)
 
**Week 2: Advanced Control Flow & Logic**
 
Topics covered: problem decomposition, abstraction, pattern recognition, algorithm design, combining loops/conditionals/strings/math into complete solutions. This is the integrative capstone for Week 2.

### Q1. Decomposition — Breaking Down a Problem
A vending machine accepts coins (1, 5, 10, 25 paisa) and sells items at various prices. Break this problem into sub-functions before writing any logic: `accept_coins(coins)`, `calculate_change(paid, price)`, `dispense_change(amount)` (greedy coin selection), and `process_purchase(price, coins)`. Write each function, then wire them together. Test with price=37, coins=[25, 25].

### Q2. Pattern Recognition — Detecting Number Sequences
Given a list of numbers, write a function `identify_sequence(nums)` that detects and returns:
- `"arithmetic"` if the difference between consecutive terms is constant
- `"geometric"` if the ratio between consecutive terms is constant
- `"fibonacci-like"` if each term equals the sum of the two previous
- `"unknown"` otherwise
Test with `[2, 4, 6, 8]`, `[3, 6, 12, 24]`, `[1, 1, 2, 3, 5, 8]`, `[1, 4, 9, 16]`.
 

### Q3. Abstraction — Generic Sorting Key Builder
Instead of writing a new sort key for every use case, write a function `make_sort_key(*attributes)` that returns a key function extracting the given attributes (by name if dicts, by index if tuples). Use it to sort a list of employee dicts first by `department`, then by `salary` descending — by passing `make_sort_key("department", "salary")`.

### Q4. Algorithm Design — Greedy Scheduling
You have a list of tasks, each with a `start` and `end` time. Find the maximum number of non-overlapping tasks you can schedule (Activity Selection Problem). Use a greedy algorithm: always pick the task that ends earliest and doesn't overlap the last chosen task. Write the function, test it with at least 8 tasks, and print the selected schedule.

### Q5. Combining Strings and Math — Roman Numerals
Write `to_roman(n)` converting an integer (1–3999) to a Roman numeral string, and `from_roman(s)` converting back to an integer. Use a lookup table of value-symbol pairs in descending order. Verify round-trip: `from_roman(to_roman(n)) == n` for all integers 1–3999.

### Q6. Loops + Conditionals — Simulation
Simulate a simple bank with multiple accounts. Process a list of transactions (each a dict with `account_id`, `type` (`"deposit"`/`"withdraw"`), and `amount`). Use a dictionary of balances. Print a final statement for each account. Reject withdrawals that would cause a negative balance (log them separately). Test with at least 3 accounts and 10 transactions.

### Q7. Recursion + Pattern — Tower of Hanoi
Implement the Tower of Hanoi algorithm for `n` disks and 3 pegs (A, B, C). Print each move as `"Move disk N from X to Y"`. Count total moves and verify it equals `2^n - 1`. Then write an iterative version using a stack. Test for `n = 3` and `n = 4`.

### Q8. String + Logic — Text Justification
Write a function `justify_text(text, width)` that formats a block of text so each line is exactly `width` characters wide using full justification (spaces distributed evenly between words). The last line of each paragraph should be left-aligned. Test with a paragraph of at least 4 lines at width=40.

### Q9. Math + Loop — Numerical Methods
Implement Newton's method to find the square root of a number `n` without using `math.sqrt`:
- Start with initial guess `x = n / 2`
- Iterate: `x = (x + n/x) / 2`
- Stop when `|x² - n| < tolerance` (e.g., `1e-10`)
Also implement the bisection method for finding roots of arbitrary functions. Use bisection to find the root of `f(x) = x³ - x - 2` on the interval `[1, 2]`.

### Q10. Abstraction + Data — CSV Report Generator
Given raw data (a list of dicts representing sales), write a `generate_report(data, group_by, metrics)` function that:
- Groups data by any specified field
- Computes any combination of `"sum"`, `"mean"`, `"count"`, `"max"`, `"min"` for specified numeric fields
- Returns the result as a formatted string table with aligned columns
This is a simplified reimplementation of what `pandas.groupby().agg()` does — build it from scratch using only dicts and loops.

### Q11. Pattern Recognition + Math — Magic Square Validator
A magic square is an `n×n` grid where all rows, columns, and both diagonals sum to the same value. Write `is_magic_square(matrix)` that validates this. Then write `generate_magic_square(n)` for odd `n` using the Siamese method algorithm. Verify generated squares pass validation for `n = 3, 5, 7`.

### Q12. Decomposition + Strings — Mini Markdown Parser
Write a `parse_markdown(text)` function that converts a small subset of Markdown to plain-text with formatting hints:
- `# Heading` → `=== Heading ===`
- `**bold**` → `[BOLD: bold]`
- `*italic*` → `[ITALIC: italic]`
- `- item` → `  • item`
- Blank lines → preserved as paragraph breaks
Process a sample Markdown document of at least 10 lines.
 

### Q13. Algorithm Design — Scheduling with Priorities
Extend Q4's scheduling problem: each task now has a `priority` (1–5) in addition to `start`/`end`. Modify the algorithm to: first schedule all priority-5 tasks (greedy by end time), then fill remaining slots with lower-priority tasks. If two tasks of the same priority conflict, prefer the one ending earlier. Test with 12 tasks of mixed priorities.

### Q14. Combining Everything — Word Frequency Analyzer
Write a complete `WordFrequencyAnalyzer` that accepts a large text and computes:
- Top 10 most frequent words (excluding stopwords like "the", "a", "is", "in", etc.)
- Average word frequency
- Words that appear exactly once ("hapax legomena")
- Frequency distribution binned into: rare (1–2), occasional (3–10), frequent (>10)
- A simple bar chart of top 10 words printed as ASCII bars (`█` characters scaled to frequency)
Test on at least 200 words of text.

### Q15. Capstone Challenge — Mini Interpreter
Build a tiny interpreter for a simple expression language. It should handle:
- Integer literals and float literals
- Arithmetic: `+`, `-`, `*`, `/`, `//`, `%`, `**`
- Parentheses for grouping
- Variables: `x = 5` assigns 5 to `x`; subsequent expressions can reference `x`
- A `print(expr)` command
Implement it as a REPL that processes a list of command strings (simulating user input). Use recursion or a stack for expression parsing. Test with:
```
x = 10
y = 3
print(x * y + 2)
print((x + y) ** 2)
z = x // y
print(z % 2)
```
This integrates: string parsing, conditionals, loops, math, recursion, and data structures — everything from Weeks 1 and 2.
