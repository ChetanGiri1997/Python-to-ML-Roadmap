### Q15. Challenge — Profiling and Optimizing a Real Function
Given this slow function that checks if any two numbers in a list sum to a target:
```python
def two_sum_slow(numbers, target):
    for i in range(len(numbers)):
        for j in range(len(numbers)):
            if i != j and numbers[i] + numbers[j] == target:
                return (i, j)
    return None
```
