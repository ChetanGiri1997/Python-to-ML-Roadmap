```python
def compute(x):
    return x ** 2 + 2 * x + 1


if __name__ == "__main__":
    print(compute(3))   # 16
    print(compute(10))  # 121

```

### why this guard exists:
- when Python runs a file directly, __name__ is set to "__main__"
- when the file is imported as a module, __name__ is set to the module's filename
-
- without this guard, compute() would execute on every import — unintended side effects
- with it:
-   - `python script.py`       → __name__ == "__main__" → block runs
-   - `import script`          → __name__ == "script"   → block skipped
-
- use it to separate reusable library code from executable entry-point logic


### Q9. `collections.Counter`
Given a long string of text, use `collections.Counter` to find the 5 most common words. (Split the text into words first.) Demonstrate `most_common(n)`, arithmetic between counters, and the `elements()` method.
 

```python
from collections import Counter

text = "the quick brown fox jumps over the lazy dog the fox and the dog slept the dog barked"
words = text.split()

c = Counter(words)

print(c.most_common(5))  # [('the', 5), ('dog', 3), ('fox', 2), ...]

c2 = Counter({"the": 2, "fox": 1, "cat": 4})

print(c + c2)   # combine counts
print(c - c2)   # subtract (drops zero/negative)
print(c & c2)   # intersection: min of each
print(c | c2)   # union: max of each

print(list(Counter({"a": 3, "b": 2}).elements()))  # ['a', 'a', 'a', 'b', 'b']
```

### Q10. `collections.defaultdict`
Use `collections.defaultdict(list)` to group a list of `(name, city)` tuples by city, building a dictionary where each key is a city and the value is a list of names in that city.

```python
from collections import defaultdict

data = [
    ("Alice", "NYC"), ("Bob", "LA"), ("Charlie", "NYC"),
    ("Diana", "LA"), ("Eve", "Chicago"), ("Frank", "NYC"),
]

city_map = defaultdict(list)
for name, city in data:
    city_map[city].append(name)

print(dict(city_map))
# {'NYC': ['Alice', 'Charlie', 'Frank'], 'LA': ['Bob', 'Diana'], 'Chicago': ['Eve']}
```

Create a named tuple `Employee` with fields `name`, `department`, `salary`. Create at least 4 instances and demonstrate: accessing by name, accessing by index, using `._asdict()`, and creating a modified copy with `._replace()`.

### Q12. `collections.OrderedDict`
Demonstrate a practical use case for `collections.OrderedDict` — create one from a list of key-value pairs, show that insertion order is maintained, and demonstrate `move_to_end()`.

```python
from collections import OrderedDict

# practical use: LRU-style cache maintaining access order
cache = OrderedDict([
    ("user:1", "Alice"),
    ("user:2", "Bob"),
    ("user:3", "Charlie"),
])

print(list(cache.keys()))  # ['user:1', 'user:2', 'user:3'] — insertion order preserved

# simulate cache access: move accessed item to end (most recently used)
cache.move_to_end("user:1")
print(list(cache.keys()))  # ['user:2', 'user:3', 'user:1']

# move to front (least recently used)
cache.move_to_end("user:3", last=False)
print(list(cache.keys()))  # ['user:3', 'user:2', 'user:1']

# evict LRU (first item)
evicted = cache.popitem(last=False)
print(f"Evicted: {evicted}")       # Evicted: ('user:3', 'Charlie')
print(dict(cache))                 # {'user:2': 'Bob', 'user:1': 'Alice'}
```

### Q13. `collections.deque`
Use `collections.deque` with a `maxlen` of 5 to implement a "recent history" log: append 10 items one by one and show that only the last 5 are retained. Also demonstrate `appendleft()`, `rotate()`, and `popleft()`.
 

```python
from collections import deque

history = deque(maxlen=5)

for i in range(1, 11):
    history.append(f"event_{i}")

print(history)  # deque(['event_6', 'event_7', 'event_8', 'event_9', 'event_10'], maxlen=5)

history.appendleft("URGENT")
print(history)  # deque(['URGENT', 'event_6', 'event_7', 'event_8', 'event_9'], maxlen=5)

history.rotate(2)   # shift right by 2
print(history)  # deque(['event_8', 'event_9', 'URGENT', 'event_6', 'event_7'], maxlen=5)

history.rotate(-2)  # shift left by 2 (undo)
print(history)

print(history.popleft())  # URGENT
print(history)  # deque(['event_6', 'event_7', 'event_8', 'event_9'], maxlen=5)
```

### Q14. `itertools` Deep Dive
Using `itertools`, demonstrate the following with a small example each:
- `chain` — flattening multiple iterables
- `product` — Cartesian product (e.g., all combinations of two lists)
- `groupby` — grouping sorted data
- `accumulate` — running total/cumulative sum
- `cycle` — cycling through a sequence a fixed number of times (use `islice` to limit)

```python
import itertools

# chain — flatten multiple iterables
print(list(itertools.chain([1, 2], [3, 4], [5, 6])))
# [1, 2, 3, 4, 5, 6]

# product — cartesian product
print(list(itertools.product(["S", "M", "L"], ["red", "blue"])))
# [('S', 'red'), ('S', 'blue'), ('M', 'red'), ('M', 'blue'), ('L', 'red'), ('L', 'blue')]

# groupby — group consecutive sorted data
data = sorted([("fruit", "apple"), ("veggie", "carrot"), ("fruit", "banana"), ("veggie", "pea")], key=lambda x: x[0])
for key, group in itertools.groupby(data, key=lambda x: x[0]):
    print(key, [item[1] for item in group])
# fruit  ['apple', 'banana']
# veggie ['carrot', 'pea']

# accumulate — running total
print(list(itertools.accumulate([1, 2, 3, 4, 5])))
# [1, 3, 6, 10, 15]

# cycle — repeat sequence, limited with islice
print(list(itertools.islice(itertools.cycle(["a", "b", "c"]), 8)))
# ['a', 'b', 'c', 'a', 'b', 'c', 'a', 'b']
```

### Q20. Lazy Imports and Import Performance
Write a module where a slow, heavy import (simulate with `time.sleep(0.5)`) is done inside a function rather than at the top level. Show that importing the module is fast, but calling the function that needs it will be slower the first time. Discuss in a comment the trade-off between top-level and lazy imports.

```python
import time

# --- simulate top-level (eager) import ---
start = time.perf_counter()
import json                      # fast standard lib, but imagine this takes 0.5s
print(f"eager import : {time.perf_counter() - start:.4f}s")

# --- lazy import pattern ---
with open("heavy_module.py", "w") as f:
    f.write('''
_heavy = None

def get_heavy():
    global _heavy
    if _heavy is None:
        import time
        time.sleep(0.5)          # simulate slow import (e.g. numpy, tensorflow)
        _heavy = {"loaded": True}
    return _heavy

def cheap_func():
    return "no heavy import needed"
''')

```

```python
import sys
sys.modules.pop("heavy_module", None)
import heavy_module
```

```python
start = time.perf_counter()
print(heavy_module.cheap_func())              # instant — no sleep triggered
print(f"cheap_func   : {time.perf_counter() - start:.4f}s")
```

```python
start = time.perf_counter()
print(heavy_module.get_heavy())              # slow — triggers sleep on first call
print(f"get_heavy #1 : {time.perf_counter() - start:.4f}s")
```

```python
start = time.perf_counter()
print(heavy_module.get_heavy())              # fast — cached in _heavy
print(f"get_heavy #2 : {time.perf_counter() - start:.4f}s")
```

### trade-offs:
- top-level imports  — fail fast (ImportError at startup), always available,
-                      easier to read, but slow down module load unconditionally
- lazy imports       — faster startup, useful when import is rarely/conditionally needed (e.g. optional dependency, CLI subcommand, plugin)
-                     downside: errors surface late, harder to spot missing deps

### Q22. Inspecting a Module with `inspect`
Import the `inspect` module and use it to:

```python
import inspect, math

# all functions in math module
math_funcs = [name for name, obj in inspect.getmembers(math, inspect.isbuiltin)]
print(math_funcs)

# getsource on a user-defined function
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(inspect.getsource(greet))

# signature of user-defined functions only — builtins lack introspectable signatures
print(inspect.signature(greet))   # (name, greeting='Hello')

def power(base, exp=2):
    return base ** exp

print(inspect.signature(power))   # (base, exp=2)

for name, param in inspect.signature(greet).parameters.items():
    print(f"{name}: default={param.default}")
# name: default=<class 'inspect._empty'>
# greeting: default=Hello

# inspect.signature() raises ValueError on C-extension builtins (math.log, len, etc.)
# because they have no Python-level signature metadata — use help() or docs instead
```

### Q24. Challenge — Build a Mini Plugin System
Design a simple plugin system using modules: write a `plugin_manager.py` that maintains a registry (a dictionary) of named "plugins". Create at least two plugin modules (`plugin_a.py`, `plugin_b.py`) each exposing a `run()` function. The plugin manager should be able to load a plugin by name (using `importlib.import_module`), register it, list all registered plugins, and execute any registered plugin's `run()` by name. Demonstrate loading and running both plugins through the manager.

```python
import os, sys, importlib

# --- write plugin files ---
with open("plugin_a.py", "w") as f:
    f.write('def run(): return "Plugin A: normalizing data..."\n')

with open("plugin_b.py", "w") as f:
    f.write('def run(): return "Plugin B: generating report..."\n')

# --- plugin_manager.py ---
with open("plugin_manager.py", "w") as f:
    f.write('''
import importlib

_registry = {}

def load(name):
    mod = importlib.import_module(name)
    if not hasattr(mod, "run"):
        raise AttributeError(f"{name} has no run()")
    _registry[name] = mod

def list_plugins():
    return list(_registry.keys())

def execute(name):
    if name not in _registry:
        raise KeyError(f"Plugin '{name}' not registered")
    return _registry[name].run()
''')

# clear stale cache
for mod in ["plugin_a", "plugin_b", "plugin_manager"]:
    sys.modules.pop(mod, None)

import plugin_manager as pm

pm.load("plugin_a")
pm.load("plugin_b")

print(pm.list_plugins())        # ['plugin_a', 'plugin_b']
print(pm.execute("plugin_a"))   # Plugin A: normalizing data...
print(pm.execute("plugin_b"))   # Plugin B: generating report...

try:
    pm.execute("plugin_c")
except KeyError as e:
    print(e)                    # "Plugin 'plugin_c' not registered"
```
