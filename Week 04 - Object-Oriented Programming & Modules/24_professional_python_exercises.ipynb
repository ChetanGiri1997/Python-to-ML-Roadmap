# Day 24 — Professional Python (Week 4 Capstone)
 
**Week 4: Object-Oriented Programming & Modules**
 
This is the integrative capstone for Week 4. Problems combine OOP, design patterns, modules, error handling, and professional Python practices: type hints, docstrings, testing with `unittest`/`pytest`, code organization, performance considerations, and Pythonic idioms.

### Q1. Full Type Hint Coverage
Take any class you wrote this week (at least 5 methods) and add complete type hints to every method signature and attribute. Use `Optional`, `List`, `Dict`, `Tuple` from `typing` (or built-in generics for Python 3.9+). Run `mypy` on it (or explain how you'd do so in a comment) and fix any type errors.
 

### Q2. Comprehensive Docstrings (NumPy/Google Style)
Write a class `StatisticsCalculator` with methods `mean`, `median`, `variance`, and `std_dev`, each with a complete NumPy-style or Google-style docstring (parameters, returns, raises, examples). Use `help()` to verify the output looks professional.

### Q3. Unit Testing with `unittest`
Write a `unittest.TestCase` class for the `StatisticsCalculator` from Q2. Include test methods for: normal inputs, empty list (should raise), single-element list, all-same-elements list, and negative numbers. Use `assertEqual`, `assertAlmostEqual`, and `assertRaises`.


### Q4. Parameterized Tests
Using `unittest` and `subTest()` (or `pytest.mark.parametrize` if using pytest), write a single test method that runs the same assertion logic against a table of (input, expected_output) pairs, avoiding code duplication across test cases.

### Q5. Test Fixtures and `setUp`/`tearDown`
Write a test class for a `FileProcessor` (reads/writes files) using `setUp` to create temporary test files and `tearDown` to clean them up. Use Python's built-in `tempfile` module for safe temporary file creation.
 

### Q6. Mocking with `unittest.mock`
Write a class `WeatherService` that makes an HTTP call (simulate with a function that returns fake data or raises). In your test, use `unittest.mock.patch` to mock the HTTP call so tests don't actually hit the network. Verify the mock was called with the right arguments.

### Q7. Property-Based Testing Concept (Hypothesis)
Without necessarily installing `hypothesis`, write a function `reverse_words(s)` and describe (in detailed comments) 3 properties that should always hold (e.g., reversing twice returns the original), then write normal `unittest` tests that check those properties against a range of manually chosen inputs covering edge cases.

### Q8. Professional `__repr__` and `__str__` Contract
Write a class `Vector3D` where `repr(v)` produces a string that, when passed to `eval()`, recreates the same vector (i.e., `eval(repr(v)) == v`). Verify this contract with a test.


### Q9. Implementing a Protocol (Structural Subtyping)
Using `typing.Protocol`, define a `Drawable` protocol with a `draw() -> str` method. Write a function `render_all(items: list[Drawable]) -> list[str]` that calls `draw()` on each item. Create three unrelated classes that satisfy the protocol without explicitly inheriting from it. Verify with `isinstance(obj, Drawable)` using `runtime_checkable`.

### Q10. `__slots__` with Inheritance
Create a base class `Base` with `__slots__ = ('x',)` and a subclass `Derived(Base)` with `__slots__ = ('y',)`. Show that instances of `Derived` have both `x` and `y` as slotted attributes. Show what happens if you forget `__slots__` in the subclass (a `__dict__` reappears).

### Q11. Descriptor Protocol
Implement a `TypeChecked` descriptor that, when used as a class attribute, enforces that the assigned value is of a specified type. Use it to create a `Person` class where `name` must be `str` and `age` must be `int` — any wrong-type assignment raises `TypeError`.

### Q12. `__init_subclass__` for Enforcing Subclass Contracts
Write a base class `APIEndpoint` that uses `__init_subclass__` to require all subclasses to define a class attribute `path` (string) and a class attribute `method` (one of `"GET"`, `"POST"`, `"PUT"`, `"DELETE"`). If a subclass is defined without these, raise `TypeError` immediately at class definition time (not at instantiation time).

### Q13. Context Manager Protocol in OOP
Write a class `Timer` that works as a context manager: `__enter__` records the start time and returns `self`, `__exit__` records the stop time. The `Timer` object then has `.elapsed` attribute available after the `with` block. Test it by timing a small computation. Write a test verifying `.elapsed > 0`.

### Q14. Operator Overloading for a Practical Class
Create a `Money` class with `amount` (Decimal for precision) and `currency` (str). Implement `__add__`, `__sub__`, `__mul__` (scalar), `__truediv__` (scalar), `__eq__`, `__lt__`, and `__str__`. Raise `TypeError` when adding money of different currencies. Write tests for all operators.

### Q15. Profiling Code with `cProfile`
Write two versions of a function that finds all prime numbers up to `n` (one naive, one with the Sieve of Eratosthenes). Use `cProfile.run()` to profile both for `n = 100000`. Print a comparison of total time and number of function calls.

### Q16. Memory Profiling with `tracemalloc`
Use Python's built-in `tracemalloc` module to measure the peak memory usage of: (a) storing 1 million integers in a plain list, (b) in a `numpy` array (if available), and (c) in a generator that computes them on-the-fly. Print a memory usage comparison.


### Q17. Dataclasses — Advanced Features
Create a `@dataclass` called `Config` with: a regular field, a field with a default factory (`field(default_factory=list)`), a field excluded from `__repr__` (`repr=False`), a field excluded from comparison (`compare=False`), and a `__post_init__` that validates/transforms the data. Demonstrate `frozen=True` for an immutable dataclass.

### Q18. `enum` Module for Typed Constants
Refactor a class that uses string literals for states (`"pending"`, `"active"`, `"closed"`) to use a proper `enum.Enum`. Add `@property` methods to the enum for extra behavior (e.g., `is_terminal`). Show how enums prevent typo bugs that raw strings allow.

### Q19. `abc` + `dataclasses` Combined
Create an abstract dataclass (using both `@dataclass` and `abc.ABC`) representing a `Transaction`. Define abstract methods `process()` and `validate()`, and concrete dataclass fields `amount` and `timestamp`. Create a concrete subclass `PaymentTransaction` that implements the abstract methods.

### Q20. Packaging — `setup.py` / `pyproject.toml` Structure
Write a complete (but minimal) `pyproject.toml` for a package you've built during Week 3–4 (use any of your classes). Include: package name, version, author, description, Python version requirement, dependencies, and entry points. Explain in comments each section's purpose.
 

### Q21. Code Style and Linting
Take a deliberately poorly styled function (inconsistent naming, long lines, missing spaces, shadowed builtins) and rewrite it to fully comply with PEP 8. List the specific PEP 8 rules you applied. Mention which tools (`flake8`, `black`, `isort`, `pylint`) would catch/fix which issues.

### Q22. `functools.total_ordering`
Create a `Temperature` class with only `__eq__` and `__lt__` implemented. Use `@functools.total_ordering` to automatically derive `__le__`, `__gt__`, and `__ge__`. Test all 6 comparison operators.

### Q23. Thread Safety with OOP
Write a `SafeCounter` class that is safe to use from multiple threads simultaneously, using `threading.Lock` inside the class (acquired in `increment()` and `get()`). Write a test that creates 10 threads each incrementing the counter 1000 times, then verifies the final count is exactly 10,000.

### Q24. Real-World Mini Project — Configuration Management System
Build a `ConfigManager` class with the following features:
- Loads settings from a JSON file at construction
- Provides `get(key, default=None)` and `set(key, value)` methods
- Supports dotted key access: `get("database.host")` navigates nested dicts
- Validates config against a schema (at minimum, checks required keys exist)
- Saves changes back to the JSON file with `save()`
- Uses `@property` for a `is_dirty` flag tracking unsaved changes
- Custom exceptions: `ConfigKeyError`, `ConfigValidationError`
Write at least 6 unit tests covering the main behaviors.

### Q25. Capstone Challenge — Professional Library Package
Build a complete, professional mini-library called `datapipe` that demonstrates everything from weeks 3–4:
 
**Structure:**
```
datapipe/
├── __init__.py
├── pipeline.py       # Core Pipeline class (Builder pattern)
├── steps.py          # Abstract Step class + concrete steps (Filter, Map, Reduce)
├── storage.py        # Abstract DataStore + FileStore, MemoryStore
├── exceptions.py     # Custom exception hierarchy
└── utils.py          # Decorators (timer, logger), validators
```
 
**Requirements:**
1. `Pipeline` is built with a `Builder` — steps are added via chaining: `Pipeline.builder().add_step(FilterStep(...)).add_step(MapStep(...)).build()`
2. Steps use the `Strategy` pattern — each has an abstract `process(data)` method
3. `DataStore` is an abstract base class; `FileStore` reads/writes JSON; `MemoryStore` uses a dict
4. All errors use a custom exception hierarchy (`DataPipeError` → `ValidationError`, `StepError`, `StorageError`)
5. A `@timer` decorator from `utils.py` applied to `Pipeline.run()` reports execution time
6. All public classes and functions have proper docstrings and type hints
7. A `tests/` folder with at least 10 `unittest` test cases
Write the complete code for all files and run the pipeline on sample data (a list of dicts) demonstrating filter, transform, and aggregation.
