# Day 23 — Design Patterns
 
**Week 4: Object-Oriented Programming & Modules**
 
Topics covered: Creational patterns (Singleton, Factory, Builder), Structural patterns (Adapter, Decorator, Facade, Proxy), Behavioral patterns (Observer, Strategy, Command, Iterator, State), and Pythonic alternatives to classic patterns.
 
---

### Q1. Singleton — Classic Implementation
Implement a `Singleton` class where only one instance can ever be created. Override `__new__` to return the existing instance on subsequent calls. Verify that `Singleton() is Singleton()` is `True`.

### Q2. Singleton — Metaclass Version
Reimplement Singleton using a metaclass `SingletonMeta` that overrides `__call__`. Apply it to a `Logger` class. Show that the same instance is returned regardless of how many times `Logger()` is called.

### Q3. Singleton — Module-Level (Pythonic)
Explain and demonstrate the most Pythonic Singleton: simply using a module-level variable. Write a module `config.py` (create it on disk) with a single `Config` object. Import it from two different places and show they reference the same object.

### Q4. Simple Factory Function
Write a factory function `create_animal(animal_type, name)` that returns an instance of `Dog`, `Cat`, or `Bird` based on the `animal_type` string, raising `ValueError` for unknown types. This avoids the caller needing to know which class to import.

### Q5. Factory Method Pattern
Create an abstract `Notification` class with an abstract factory method `create_message()`. Subclass it as `EmailNotification` and `SMSNotification`, each implementing `create_message()` to return a different message-type object. The parent class has a `send()` method that calls `create_message()` — it doesn't know what type it creates.

### Q6. Abstract Factory Pattern
Create two abstract factories `LightThemeFactory` and `DarkThemeFactory`, each producing matching `Button` and `TextBox` objects styled for their theme. Write client code that works exclusively through the abstract factory interface — no `if/else` for theme selection in the client.
 

### Q7. Builder Pattern
Design a `PizzaBuilder` that lets you configure a pizza step by step: `set_size()`, `add_topping()`, `set_crust()`, `set_sauce()`, with a final `build()` returning a `Pizza` object. Each setter should return `self` for chaining. Build two different pizzas using the same builder class.


### Q8. Builder with a Director
Add a `PizzaDirector` class that knows how to build specific pre-configured pizzas (e.g., `make_margherita(builder)`, `make_bbq_chicken(builder)`) by calling the builder's methods in the right sequence. This separates the construction sequence from the builder's individual steps.
 

### Q9. Prototype Pattern
Implement a `Prototype` mixin with a `clone()` method that returns a deep copy of the object. Apply it to a `Config` class with nested settings. Demonstrate modifying the clone doesn't affect the original.
 

### Q10. Adapter Pattern
You have a legacy `OldCalculator` class with a method `do_operation(op_string, a, b)` (e.g., `"add"`, `"mul"`). You need it to work with code expecting a `Calculator` interface with methods `add(a, b)`, `subtract(a, b)`, `multiply(a, b)`. Write an `Adapter` class that wraps `OldCalculator` and exposes the new interface.

### Q11. Decorator Pattern (Structural, Not Python `@decorator`)
Create a `Coffee` base class with `cost()` and `description()`. Create decorator classes `MilkDecorator`, `SugarDecorator`, and `VanillaDecorator` that each wrap a coffee object and add to its cost and description. Demonstrate stacking multiple decorators: `VanillaDecorator(MilkDecorator(Coffee()))`.

### Q12. Facade Pattern
Create a facade class `HomeTheaterFacade` that hides the complexity of operating several subsystems: `DVDPlayer`, `Projector`, `SurroundSound`, and `Lights`. The facade provides simple methods `watch_movie(title)` and `end_movie()` that internally coordinate the subsystems in the right order.

### Q13. Proxy Pattern — Virtual Proxy (Lazy Loading)
Create a `HeavyResource` class that simulates expensive initialization (use `time.sleep(1)` in `__init__`). Create a `LazyProxy` class that only initializes the real object when a method is first called on it, not at proxy construction time. Demonstrate the difference in startup time.
 

### Q14. Proxy Pattern — Protection Proxy
Create a `SecureFileAccess` proxy around a `FileReader` class. The proxy checks if the current user (passed at construction) has read permission before delegating to the real `FileReader`. Unauthorized access raises `PermissionError`.

### Q15. Observer Pattern
Implement an `EventEmitter` class (Subject) with `subscribe(event, callback)`, `unsubscribe(event, callback)`, and `emit(event, *args)` methods. Create three observer functions. Subscribe them to events, emit events, and demonstrate that unsubscribed observers no longer receive notifications.

### Q16. Strategy Pattern
Create a `Sorter` class that accepts a sorting strategy function at construction time. Implement three strategies: `bubble_sort`, `python_builtin_sort`, and `reverse_sort` (all taking a list and returning a sorted list). Swap strategies at runtime and show the same data sorted differently.

### Q17. Strategy Pattern — With Abstract Class
Reimplement Q16 using an abstract `SortStrategy` base class with an abstract `sort(data)` method, making the strategies full classes rather than plain functions. Discuss in a comment when class-based strategies are preferred over function strategies.

### Q18. Command Pattern
Build a simple text editor with undo/redo support using the Command pattern. Each command (`InsertTextCommand`, `DeleteTextCommand`) stores enough state to `execute()` and `undo()` itself. A `TextEditor` class holds a list of executed commands (as a stack) and supports `undo_last()` and `redo_last()`.

### Q19. Iterator Pattern
Implement a custom `TreeNode` class for a binary tree, and write a separate `InOrderIterator` class that iterates over the tree's nodes in in-order sequence, implementing `__iter__` and `__next__`. Demonstrate iterating over the tree with a `for` loop.

### Q20. State Pattern
Model a `TrafficLight` using the State pattern. Create state classes `RedState`, `YellowState`, and `GreenState` — each implementing a `next_state()` method that transitions the light to the next state. The `TrafficLight` class delegates behavior to its current state. Cycle through 6 transitions and print each state.

### Q21. Chain of Responsibility
Build a support ticket system where a ticket with a `severity` level (1–5) is passed along a chain of handlers: `JuniorSupport` handles severity 1–2, `SeniorSupport` handles 3–4, `ManagementEscalation` handles 5. If a handler can't handle it, it passes to the next. Each handler either resolves or escalates.

### Q22. Template Method vs Strategy — Comparison
Implement the same "data export" task (load, filter, format, write) using both:
- **Template Method**: abstract base class with a concrete `export()` method calling abstract steps
- **Strategy**: a `DataExporter` class that accepts pluggable strategy objects for filtering and formatting
Write a brief comparison comment: when does each pattern win?

### Q23. Null Object Pattern
Instead of returning `None` from a function that might not find a result (forcing callers to check `if result is not None:`), create a `NullUser` class that implements the same interface as `User` but with harmless no-op methods. Show how this eliminates null checks in client code.

### Q24. Flyweight Pattern
Simulate a text editor that renders thousands of characters. Instead of creating a new `CharacterStyle` object for every character, use a `StyleFactory` that caches and reuses `CharacterStyle` objects with the same font/size/color. Show memory savings by comparing the number of unique objects created vs. the number of characters rendered.

### Q25. Challenge — Mini Event-Driven Order System
Design a small order processing system combining multiple patterns:
1. **Factory** to create different order types (`OnlineOrder`, `InStoreOrder`)
2. **Observer** to notify services when an order changes state (inventory, email, analytics)
3. **State** to model order lifecycle: `Pending → Confirmed → Shipped → Delivered` (or `Cancelled` from any state)
4. **Command** to support cancellation with undo (restore previous state)
5. **Facade** to expose a simple `OrderService.place_order(...)` and `OrderService.cancel_order(...)` API hiding all the above complexity
Build a demo that places 2 orders, ships one, cancels one (and calls undo), and prints the notification log from the observers.
