# Day 20 — Modules & Packages
 
**Week 4: Object-Oriented Programming & Modules**
 
Topics covered: `import` mechanics, `from ... import`, aliasing, `__name__`, creating your own modules/packages, `__init__.py`, relative imports, `sys.path`, standard library deep dive (`os`, `sys`, `pathlib`, `datetime`, `collections`, `itertools`, `math`, `random`).
 

### Q1. Importing a Standard Library Module
Import the `math` module and use at least 5 different functions/constants from it (`sqrt`, `pi`, `floor`, `ceil`, `log`). Print the results with descriptive labels.

### Q2. `from ... import` vs `import`
Demonstrate the difference between `import random` and `from random import randint, choice`. Write equivalent code using each style and explain in a comment when you'd prefer one over the other.

### Q3. Aliasing Imports
Import `datetime` as `dt` and `collections.defaultdict` as `dd`. Use both with their aliases. Explain in a comment when aliasing is appropriate.

### Q4. `__name__ == "__main__"` Guard
Write a Python script (simulate it as a code cell) with a function `compute(x)` and a block `if __name__ == "__main__": ...` that calls `compute`. Explain in a comment why this guard exists and what happens when the module is imported vs run directly.

### Q5. Exploring a Module with `dir()` and `help()`
Use `dir(os)` to list all attributes/functions of the `os` module. Then use `help(os.path.join)` to display its documentation. Identify and call 5 useful functions from `os.path`.

### Q6. `os` Module — Path Operations
Using the `os` module, write code that:

- Joins path components safely (cross-platform)
- Checks if a path exists
- Lists all files in the current directory
- Gets the file extension of a path string


### Q7. `pathlib.Path` — Modern Path Handling
Rewrite the tasks from Q6 using `pathlib.Path` instead of `os`/`os.path`. Demonstrate how `Path` objects support `/` for path joining and how they're more readable than `os.path` equivalents.

### Q8. `datetime` — Working with Dates and Times

### Q9. `collections.Counter`

### Q10. `collections.defaultdict`

### Q11. `collections.namedtuple`

### Q12. `collections.OrderedDict`

### Q13. `collections.deque`

### Q18. Creating a Package (`__init__.py`)

### Q19. Relative Imports (Conceptual)

Explain when absolute vs relative imports are preferred.

### Q20. `__all__` in a Module
Write a module that defines 5 functions but sets `__all__` to expose only 3 of them. Demonstrate that `from mymodule import *` only imports the 3 listed in `__all__`, and that the other 2 can still be accessed explicitly.

### Q21. Lazy Imports and Import Performance
Write a module where a slow, heavy import (simulate with `time.sleep(0.5)`) is done inside a function rather than at the top level. Show that importing the module is fast, but calling the function that needs it will be slower the first time. Discuss in a comment the trade-off between top-level and lazy imports.
