# Day 19 — OOP Fundamentals
 
**Week 4: Object-Oriented Programming & Modules**
 
Topics covered: classes, `__init__`, instance attributes, instance methods, `self`, class attributes, `__str__`/`__repr__`, basic encapsulation.

### Q1. Your First Class
Define a class `Dog` with an `__init__` method that accepts `name` and `breed`. Create two `Dog` instances and print their attributes directly.

### Q2. Instance Methods
Add a method `bark()` to the `Dog` class that returns the string `"{name} says: Woof!"`. Call it on both instances from Q1.

### Q3. `__str__` vs `__repr__`
Add both `__str__` and `__repr__` methods to the `Dog` class. `__str__` should return a human-friendly string; `__repr__` should return a developer-friendly string (e.g., `Dog(name='Rex', breed='Labrador')`). Demonstrate both by using `print()` and `repr()`.

### Q4. Class Attribute
Add a class attribute `species = "Canis lupus familiaris"` to the `Dog` class. Access it via the class itself and via an instance. Then change it on one instance (instance attribute override) and show that the class attribute remains unchanged for other instances.

### Q5. Counting Instances
Add a class attribute `count = 0` to a class `Car`. Increment it inside `__init__` every time a new `Car` is created. Write a class method `get_count()` decorated with `@classmethod` that returns the current count. Create 4 `Car` objects and verify the count.

### Q6. Static Method
Add a static method `is_valid_speed(speed)` to the `Car` class (from Q5) that returns `True` if `speed` is between 0 and 300, `False` otherwise. Call it without creating a `Car` instance.

### Q7. Encapsulation with Name Mangling
Create a class `BankAccount` with a private attribute `__balance` (double underscore). Write `deposit(amount)` and `withdraw(amount)` methods that modify `__balance`. Prevent withdrawals that would make `__balance` negative. Add a `get_balance()` method to read it. Demonstrate that directly accessing `__balance` from outside fails.

### Q8. Property Decorator
Refactor the `BankAccount` class from Q7 to expose `balance` as a `@property` instead of `get_balance()`. Also add a `@balance.setter` that raises `ValueError` if someone tries to set a negative balance directly.
 

### Q9. Method Chaining
Design a class `QueryBuilder` with methods `select(fields)`, `where(condition)`, and `limit(n)` — each should return `self` after updating internal state. This allows chaining like `QueryBuilder().select("name").where("age > 18").limit(10)`. Add a `build()` method that returns the constructed query as a string.

### Q10. `__eq__` and `__hash__`
Create a class `Point` representing a 2D point with `x` and `y` coordinates. Implement `__eq__` so two `Point` objects with the same coordinates are considered equal. Implement `__hash__` so `Point` objects can be stored in a `set`. Demonstrate with a set of points that includes duplicates.
 

### Q11. `__lt__`, `__le__`, `__gt__`, `__ge__` (Comparison Operators)
Add comparison operators to the `Point` class (compare by distance from origin). Then sort a list of `Point` objects using `sorted()` and demonstrate `min()` and `max()` work correctly.

### Q12. `__len__` and `__contains__`
Create a class `Bag` that wraps a list of items. Implement `__len__` so `len(bag)` works, and `__contains__` so the `in` keyword works on it. Add an `add(item)` method.

### Q13. `__getitem__` and `__setitem__`
Extend the `Bag` class from Q12 with `__getitem__` and `__setitem__` so items can be accessed and modified using index notation `bag[0]` and `bag[1] = "apple"`.

### Q14. `__iter__` and `__next__`
Create a class `Countdown` that takes a starting number `n` and implements `__iter__` and `__next__` so it can be used directly in a `for` loop to count down from `n` to 0, raising `StopIteration` at the end.

 

### Q15. `__call__`
Create a class `Multiplier` that takes a factor in `__init__` and implements `__call__` so instances can be used like functions. For example, `double = Multiplier(2); double(5)` should return `10`.

### Q16. Dataclass (Python 3.7+)
Rewrite the `Point` class from Q10 using `@dataclass` from the `dataclasses` module. Show that `__init__`, `__repr__`, and `__eq__` are automatically generated.

### Q17. Class with Validation in `__init__`
Create a class `Student` with attributes `name`, `age`, and `gpa`. In `__init__`, validate that `age` is between 5 and 100, and `gpa` is between 0.0 and 4.0 — raise `ValueError` with a descriptive message if not.
 

### Q18. Immutable-Style Class Using `__setattr__`
Create a class `ImmutablePoint` that overrides `__setattr__` to raise `AttributeError` if anyone tries to set an attribute after `__init__` has run. Use a flag in `__init__` to allow the initial assignment.
 

### Q19. Object Copying
Create a class `Config` with a dictionary attribute `settings`. Demonstrate the difference between a shallow copy (via `copy.copy`) and a deep copy (via `copy.deepcopy`) when the dictionary contains nested structures.
 

### Q20. `__del__` (Destructor)
Add a `__del__` method to a class `Resource` that prints a message when the object is destroyed. Create an instance, delete it with `del`, and observe when the message appears. Discuss in a comment why relying on `__del__` for cleanup is unreliable.

### Q21. Class Representing a Stack
Implement a `Stack` class backed by a list with methods `push(item)`, `pop()`, `peek()`, `is_empty()`, and `size()`. `pop()` and `peek()` should raise `IndexError` if the stack is empty. Implement `__str__` and `__len__`.

### Q22. Class Representing a Queue
Implement a `Queue` class with methods `enqueue(item)`, `dequeue()`, `front()`, `is_empty()`, and `size()`. Use `collections.deque` internally for efficient O(1) operations on both ends.

### Q23. Slots for Memory Optimization
Create a class `Pixel` with attributes `x`, `y`, and `color`. Add `__slots__ = ('x', 'y', 'color')` and demonstrate: (a) instances use less memory than without slots, (b) adding an attribute not in `__slots__` raises `AttributeError`.

### Q24. Real-World Mini Project — Library Book
Design a `Book` class with attributes `title`, `author`, `isbn`, and `available` (bool). Add methods `checkout()` and `return_book()` that change the `available` status (raising `RuntimeError` if checking out an already-checked-out book). Implement `__str__` and `__repr__`. Create a small list of books and simulate a checkout/return workflow.

### Q25. Challenge — Implement a Matrix Class
Build a `Matrix` class that wraps a 2D list. Implement:
- `__init__(data)` — store a list of lists as the matrix
- `__str__` — display it in grid format
- `__add__` — matrix addition (raise `ValueError` if dimensions don't match)
- `__mul__` — matrix multiplication (raise `ValueError` if dimensions are incompatible)
- `transpose()` — return a new `Matrix` that is the transpose
Test all operations with at least two 2x2 and one 2x3 matrix.
