```python
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds. Withdrawal denied.")

    def get_balance(self):
        return self.__balance


# Create an account
account = BankAccount(1000)

# Deposit money
account.deposit(500)
print("Balance after deposit:", account.get_balance())

# Withdraw money
account.withdraw(300)
print("Balance after withdrawal:", account.get_balance())

# Attempt to overdraw
account.withdraw(1500)
print("Balance after failed withdrawal:", account.get_balance())

# Attempt direct access to private attribute
try:
    print(account.__balance)
except AttributeError as e:
    print("Error:", e)
```

### Q8. Property Decorator
Refactor the `BankAccount` class from Q7 to expose `balance` as a `@property` instead of `get_balance()`. Also add a `@balance.setter` that raises `ValueError` if someone tries to set a negative balance directly.
 

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance

    @property
    def balance(self):
        """Getter for balance."""
        return self.__balance

    @balance.setter
    def balance(self, value):
        """Setter for balance with validation."""
        if value < 0:
            raise ValueError("Balance cannot be negative.")
        self.__balance = value

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds. Withdrawal denied.")


# Create an account
account = BankAccount(1000)

# Access balance like an attribute
print("Current balance:", account.balance)

# Deposit and withdraw
account.deposit(500)
print("After deposit:", account.balance)

account.withdraw(300)
print("After withdrawal:", account.balance)

# Set balance directly (valid)
account.balance = 2000
print("After direct update:", account.balance)

# Attempt to set a negative balance
try:
    account.balance = -100
except ValueError as e:
    print("Error:", e)
```

### Q9. Method Chaining
Design a class `QueryBuilder` with methods `select(fields)`, `where(condition)`, and `limit(n)` — each should return `self` after updating internal state. This allows chaining like `QueryBuilder().select("name").where("age > 18").limit(10)`. Add a `build()` method that returns the constructed query as a string.

```python
class QueryBuilder:
    def __init__(self):
        self._fields = "*"
        self._condition = None
        self._limit = None

    def select(self, fields):
        self._fields = fields
        return self  # Return self for chaining

    def where(self, condition):
        self._condition = condition
        return self

    def limit(self, n):
        self._limit = n
        return self

    def build(self):
        query = f"SELECT {self._fields} FROM table"

        if self._condition:
            query += f" WHERE {self._condition}"

        if self._limit is not None:
            query += f" LIMIT {self._limit}"

        return query


# Method chaining example
query = (
    QueryBuilder()
    .select("name")
    .where("age > 18")
    .limit(10)
    .build()
)

print(query)
```

### Q10. `__eq__` and `__hash__`
Create a class `Point` representing a 2D point with `x` and `y` coordinates. Implement `__eq__` so two `Point` objects with the same coordinates are considered equal. Implement `__hash__` so `Point` objects can be stored in a `set`. Demonstrate with a set of points that includes duplicates.
 

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        if not isinstance(other, Point):
            return NotImplemented
        return self.x == other.x and self.y == other.y

    def __hash__(self):
        return hash((self.x, self.y))

    def __repr__(self):
        return f"Point({self.x}, {self.y})"


# Create some points
p1 = Point(1, 2)
p2 = Point(1, 2)  # Same coordinates as p1
p3 = Point(3, 4)
p4 = Point(1, 2)  # Same coordinates as p1 and p2

# Equality check
print(p1 == p2)  # True
print(p1 == p3)  # False

# Store in a set
points = {p1, p2, p3, p4}

print(points)
print("Number of unique points:", len(points))
```

### Q11. `__lt__`, `__le__`, `__gt__`, `__ge__` (Comparison Operators)
Add comparison operators to the `Point` class (compare by distance from origin). Then sort a list of `Point` objects using `sorted()` and demonstrate `min()` and `max()` work correctly.

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_from_origin(self):
        return math.sqrt(self.x ** 2 + self.y ** 2)

    def __eq__(self, other):
        if not isinstance(other, Point):
            return NotImplemented
        return self.x == other.x and self.y == other.y

    def __hash__(self):
        return hash((self.x, self.y))

    def __lt__(self, other):
        return self.distance_from_origin() < other.distance_from_origin()

    def __le__(self, other):
        return self.distance_from_origin() <= other.distance_from_origin()

    def __gt__(self, other):
        return self.distance_from_origin() > other.distance_from_origin()

    def __ge__(self, other):
        return self.distance_from_origin() >= other.distance_from_origin()

    def __repr__(self):
        return f"Point({self.x}, {self.y})"


# Create points
points = [
    Point(3, 4),   # Distance = 5
    Point(1, 1),   # Distance ≈ 1.41
    Point(0, 2),   # Distance = 2
    Point(6, 8)    # Distance = 10
]

# Sort points by distance from origin
sorted_points = sorted(points)

print("Sorted points:", sorted_points)

# Find minimum and maximum
print("Closest point:", min(points))
print("Farthest point:", max(points))
```

### Q14. `__iter__` and `__next__`
Create a class `Countdown` that takes a starting number `n` and implements `__iter__` and `__next__` so it can be used directly in a `for` loop to count down from `n` to 0, raising `StopIteration` at the end.

 

```python
class Countdown:
    def __init__(self, n):
        self.n = n
        self.current = n

    def __iter__(self):
        self.current = self.n  # reset on each new iteration
        return self

    def __next__(self):
        if self.current < 0:
            raise StopIteration
        val = self.current
        self.current -= 1
        return val


cd = Countdown(5)
for num in cd:
    print(num)  # 5 4 3 2 1 0

# reusable
print(list(Countdown(3)))  # [3, 2, 1, 0]
```

### Q18. Immutable-Style Class Using `__setattr__`
Create a class `ImmutablePoint` that overrides `__setattr__` to raise `AttributeError` if anyone tries to set an attribute after `__init__` has run. Use a flag in `__init__` to allow the initial assignment.
 

```python
class ImmutablePoint:
    def __init__(self, x, y):
        object.__setattr__(self, '_initialized', False)
        object.__setattr__(self, 'x', x)
        object.__setattr__(self, 'y', y)
        object.__setattr__(self, '_initialized', True)

    def __setattr__(self, name, value):
        if self._initialized:
            raise AttributeError("ImmutablePoint is immutable")
        object.__setattr__(self, name, value)

    def __repr__(self):
        return f"ImmutablePoint(x={self.x}, y={self.y})"


p = ImmutablePoint(3, 4)
print(p)    # ImmutablePoint(x=3, y=4)
print(p.x)  # 3

try:
    p.x = 10
except AttributeError as e:
    print(e)  # ImmutablePoint is immutable

try:
    p.z = 5
except AttributeError as e:
    print(e)  # ImmutablePoint is immutable
```

```python
class Resource:
    def __init__(self, name):
        self.name = name
        print(f"Resource '{self.name}' acquired")

    def __del__(self):
        print(f"Resource '{self.name}' released")


r = Resource("file_handle")
del r  # triggers __del__ immediately here, but not guaranteed in general

# --- why __del__ is unreliable ---
# 1. CPython uses reference counting, so del() usually calls __del__ immediately,
#    but other implementations (PyPy, Jython) use GC and may delay or skip it.
# 2. Circular references prevent ref count from reaching 0; GC may never call __del__.
# 3. __del__ is not called at a predictable time — never rely on it for critical cleanup.
# 4. Exceptions inside __del__ are silently ignored.
# Use context managers (with / __enter__ / __exit__) for deterministic cleanup instead.
```

### Q21. Class Representing a Stack
Implement a `Stack` class backed by a list with methods `push(item)`, `pop()`, `peek()`, `is_empty()`, and `size()`. `pop()` and `peek()` should raise `IndexError` if the stack is empty. Implement `__str__` and `__len__`.

```python
class Stack:
    def __init__(self):
        self._data = []

    def push(self, item):
        self._data.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self._data.pop()

    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self._data[-1]

    def is_empty(self):
        return len(self._data) == 0

    def size(self):
        return len(self._data)

    def __len__(self):
        return len(self._data)

    def __str__(self):
        return f"Stack({self._data})"


s = Stack()
print(s.is_empty())  # True

s.push(1)
s.push(2)
s.push(3)
print(s)         # Stack([1, 2, 3])
print(len(s))    # 3
print(s.peek())  # 3
print(s.pop())   # 3
print(s)         # Stack([1, 2])

try:
    Stack().pop()
except IndexError as e:
    print(e)  # pop from empty stack
```

### Q22. Class Representing a Queue
Implement a `Queue` class with methods `enqueue(item)`, `dequeue()`, `front()`, `is_empty()`, and `size()`. Use `collections.deque` internally for efficient O(1) operations on both ends.

```python
from collections import deque

class Queue:
    def __init__(self):
        self._data = deque()

    def enqueue(self, item):
        self._data.append(item)

    def dequeue(self):
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        return self._data.popleft()

    def front(self):
        if self.is_empty():
            raise IndexError("front of empty queue")
        return self._data[0]

    def is_empty(self):
        return len(self._data) == 0

    def size(self):
        return len(self._data)

    def __len__(self):
        return len(self._data)

    def __str__(self):
        return f"Queue({list(self._data)})"


q = Queue()
print(q.is_empty())  # True

q.enqueue("a")
q.enqueue("b")
q.enqueue("c")
print(q)           # Queue(['a', 'b', 'c'])
print(q.front())   # a
print(q.dequeue()) # a
print(q)           # Queue(['b', 'c'])
print(len(q))      # 2

try:
    Queue().dequeue()
except IndexError as e:
    print(e)  # dequeue from empty queue
```

### Q23. Slots for Memory Optimization
Create a class `Pixel` with attributes `x`, `y`, and `color`. Add `__slots__ = ('x', 'y', 'color')` and demonstrate: (a) instances use less memory than without slots, (b) adding an attribute not in `__slots__` raises `AttributeError`.

```python
import sys

class Pixel:
    __slots__ = ('x', 'y', 'color')

    def __init__(self, x, y, color):
        self.x = x
        self.y = y
        self.color = color

class PixelNoSlots:
    def __init__(self, x, y, color):
        self.x = x
        self.y = y
        self.color = color


p  = Pixel(10, 20, "red")
pn = PixelNoSlots(10, 20, "red")

print(sys.getsizeof(p))               # 48  (slots version)
print(sys.getsizeof(pn))              # 48  (instance shell same...)
print(sys.getsizeof(pn.__dict__))     # 232 (but __dict__ adds overhead)
print(hasattr(p, '__dict__'))         # False — no per-instance dict

try:
    p.alpha = 255
except AttributeError as e:
    print(e)  # 'Pixel' object has no attribute 'alpha'
```

### Q24. Real-World Mini Project — Library Book
Design a `Book` class with attributes `title`, `author`, `isbn`, and `available` (bool). Add methods `checkout()` and `return_book()` that change the `available` status (raising `RuntimeError` if checking out an already-checked-out book). Implement `__str__` and `__repr__`. Create a small list of books and simulate a checkout/return workflow.

```python
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.available = True

    def checkout(self):
        if not self.available:
            raise RuntimeError(f"'{self.title}' is already checked out")
        self.available = False

    def return_book(self):
        if self.available:
            raise RuntimeError(f"'{self.title}' was not checked out")
        self.available = True

    def __str__(self):
        status = "Available" if self.available else "Checked Out"
        return f"{self.title} by {self.author} [{status}]"

    def __repr__(self):
        return f"Book(title={self.title!r}, author={self.author!r}, isbn={self.isbn!r}, available={self.available})"


library = [
    Book("1984", "George Orwell", "978-0451524935"),
    Book("Dune", "Frank Herbert", "978-0441013593"),
    Book("Clean Code", "Robert Martin", "978-0132350884"),
]

for b in library:
    print(b)
# 1984 by George Orwell [Available]
# Dune by Frank Herbert [Available]
# Clean Code by Robert Martin [Available]

library[0].checkout()
library[1].checkout()
print("\nAfter checkouts:")
for b in library:
    print(b)

library[0].return_book()
print("\nAfter return:")
print(library[0])

try:
    library[1].checkout()  # already checked out
except RuntimeError as e:
    print(e)  # 'Dune' is already checked out

print(repr(library[2]))
```

### Q25. Challenge — Implement a Matrix Class
Build a `Matrix` class that wraps a 2D list. Implement:
- `__init__(data)` — store a list of lists as the matrix
- `__str__` — display it in grid format
- `__add__` — matrix addition (raise `ValueError` if dimensions don't match)
- `__mul__` — matrix multiplication (raise `ValueError` if dimensions are incompatible)
- `transpose()` — return a new `Matrix` that is the transpose
Test all operations with at least two 2x2 and one 2x3 matrix.
 

```python
class Matrix:
    def __init__(self, data):
        self.data = data
        self.rows = len(data)
        self.cols = len(data[0])

    def __str__(self):
        return "\n".join(" ".join(f"{v:6}" for v in row) for row in self.data)

    def __add__(self, other):
        if self.rows != other.rows or self.cols != other.cols:
            raise ValueError(f"Shape mismatch: ({self.rows}x{self.cols}) + ({other.rows}x{other.cols})")
        return Matrix([
            [self.data[i][j] + other.data[i][j] for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def __mul__(self, other):
        if self.cols != other.rows:
            raise ValueError(f"Shape mismatch: ({self.rows}x{self.cols}) * ({other.rows}x{other.cols})")
        return Matrix([
            [sum(self.data[i][k] * other.data[k][j] for k in range(self.cols)) for j in range(other.cols)]
            for i in range(self.rows)
        ])

    def transpose(self):
        return Matrix([
            [self.data[i][j] for i in range(self.rows)]
            for j in range(self.cols)
        ])

    def __repr__(self):
        return f"Matrix({self.data})"


A = Matrix([[1, 2], [3, 4]])
B = Matrix([[5, 6], [7, 8]])
C = Matrix([[1, 2, 3], [4, 5, 6]])  # 2x3

print("A:");          print(A)
print("\nB:") ;        print(B)
print("\nA + B:");     print(A + B)
print("\nA * B:");     print(A * B)
print("\nTranspose C:");print(C.transpose())

try:
    A + C
except ValueError as e:
    print(f"\n{e}")  # Shape mismatch: (2x2) + (2x3)

try:
    C * A  # 2x3 * 2x2 incompatible
except ValueError as e:
    print(e)  # Shape mismatch: (2x3) * (2x2)
```
