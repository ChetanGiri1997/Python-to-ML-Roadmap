# Day 22 — Advanced OOP Patterns
 
**Week 4: Object-Oriented Programming & Modules**
 
Topics covered: inheritance, `super()`, multiple inheritance, MRO (Method Resolution Order), mixins, abstract base classes (`abc`), `@abstractmethod`, polymorphism, duck typing, `isinstance`/`issubclass`.
 
---

### Q1. Single Inheritance
Create a base class `Animal` with attributes `name` and `sound`, and a method `speak()` returning `"{name} says {sound}"`. Create subclasses `Dog` and `Cat` that override `speak()` to return their specific message. Demonstrate polymorphism by calling `speak()` on a list of mixed `Animal` objects.

### Q2. `super()` in `__init__`
Create a class `Vehicle` with `__init__(make, model, year)`. Create a subclass `ElectricVehicle` that calls `super().__init__()` and adds an extra attribute `battery_range_km`. Demonstrate that the parent's `__init__` runs correctly.

### Q3. Overriding and Extending Methods
In a class `Shape` with a `describe()` method, create subclasses `Circle` and `Rectangle` that each call `super().describe()` first (using the parent's output) and then append their own specific information. This shows extending, not fully replacing, the parent's behavior.

### Q4. `isinstance()` and `issubclass()`
Using the class hierarchy from Q1, write a function `process_animal(obj)` that uses `isinstance` to check what type of animal it is and behaves accordingly. Then use `issubclass()` to explore class relationships.
 

### Q5. Multiple Inheritance — Basic
Create three classes: `Flyable` (with `fly()` method), `Swimmable` (with `swim()` method), and `Duck` that inherits from both. Create a `Duck` instance and call both methods. Demonstrate that `Duck` is an instance of both parent classes.

### Q6. Method Resolution Order (MRO)
Create a diamond inheritance pattern: `A` → `B`, `A` → `C`, `B` + `C` → `D`. Each class defines the same method `greet()` with a different message. Print `D.__mro__` and explain the order Python uses to resolve which `greet()` is called. Show how `super()` chains correctly through the MRO.

### Q7. `super()` in Multiple Inheritance
In the diamond pattern from Q6, modify each `greet()` method to call `super().greet()` as well, so that a single call on `D` triggers all parent implementations in MRO order. Print the chain of outputs.

### Q8. Mixins
Create a `TimestampMixin` class with `created_at` and `updated_at` attributes set in an `__init__` method, and an `update_timestamp()` method. Mix it into two unrelated classes (`BlogPost` and `UserProfile`) so both get timestamping behavior without any other shared inheritance.

### Q9. Abstract Base Classes
Using `abc.ABC` and `@abstractmethod`, define an abstract class `DataStore` with abstract methods `connect()`, `read(key)`, `write(key, value)`, and `disconnect()`. Create two concrete implementations: `InMemoryStore` and a simulated `FileStore`. Show that instantiating `DataStore` directly raises `TypeError`.

### Q10. Abstract Properties
Add an `@property` decorated with `@abstractmethod` called `store_name` to the `DataStore` class. Implement it in both concrete subclasses. Show it can be accessed as a property on instances.

### Q11. Duck Typing
Write a function `make_it_quack(obj)` that simply calls `obj.quack()` without checking the type. Create three completely unrelated classes (`Duck`, `RubberDuck`, `PersonMimickingDuck`), each with a `quack()` method. Pass all three to `make_it_quack()` and show it works — then discuss in a comment how this differs from Java/C++-style interface enforcement.


### Q12. Registering Virtual Subclasses with `abc`
Using `abc.ABCMeta`, register a class that doesn't actually inherit from your ABC as a "virtual subclass" using `ABC.register(SomeClass)`. Show that `isinstance(obj, ABC)` returns `True` for instances of the registered class, even without inheritance.

### Q13. Polymorphic Area Calculator
Create an abstract `Shape` base class with an abstract method `area()` and a concrete method `describe()` that calls `area()`. Implement `Circle`, `Rectangle`, `Triangle`, and `Trapezoid` subclasses. Write a function that takes a list of mixed shapes and prints the area of each, and sorts them by area.

### Q14. Cooperative Multiple Inheritance with `super()` and `*args/**kwargs`
Show a best-practice pattern for cooperative multiple inheritance: each class in a hierarchy passes along `*args` and `**kwargs` via `super().__init__()` so that all classes in the MRO can receive their relevant keyword arguments from a single construction call. Design a 3-class example demonstrating this.

### Q15. Inheriting from Built-in Types
Create a class `CappedList` that inherits from `list` and overrides `append()` to enforce a maximum length (raising `OverflowError` if at capacity). Show that all other `list` methods (`pop`, `sort`, indexing, etc.) still work normally.

### Q16. Inheriting from `dict`
Create a `DefaultNoneDict` class inheriting from `dict` that overrides `__missing__` to return `None` (instead of raising `KeyError`) when a key doesn't exist — without having to use `collections.defaultdict`.

### Q17. `__init_subclass__`
Use `__init_subclass__` in a base class `Plugin` to automatically register all subclasses in a class-level registry dictionary as soon as they are defined. Demonstrate by creating 3 plugin subclasses and printing the registry.

### Q18. Mixins for Serialization
Write a `JSONSerializableMixin` that provides `to_json()` and a class method `from_json(json_str)` to any class that uses it. The mixin should rely on the class having a `__dict__` (or explicit fields). Apply it to two unrelated model classes and demonstrate round-trip serialization.

### Q19. Abstract Factory Pattern (OOP Version)
Create an abstract `UIFactory` class with abstract methods `create_button()` and `create_checkbox()`. Implement two concrete factories: `WindowsFactory` and `MacFactory`. Each returns platform-specific button/checkbox objects (simple classes that print their render output). Write client code that uses only the abstract factory interface.

### Q20. Template Method Pattern
Create an abstract class `DataProcessor` with a concrete method `process()` that calls abstract steps `load()`, `validate()`, `transform()`, and `save()` in sequence. Implement two concrete processors (`CSVProcessor` and `JSONProcessor`) that define their own versions of each step. Show that `process()` doesn't need to be overridden.

### Q21. Composition Over Inheritance
Refactor a deeply-nested inheritance chain (e.g., `Animal → Mammal → Pet → Dog → TrainedDog`) to use composition instead: a `Dog` class that holds references to separate capability objects (`TrainingBehavior`, `FeedingBehavior`). Discuss in comments what is gained and what is lost versus inheritance.

### Q22. `functools.singledispatch` for Function-Based Polymorphism
Use `@functools.singledispatch` to create a `serialize(obj)` function that behaves differently based on the type of `obj`: handles `int`, `float`, `str`, `list`, and `dict` each with a different implementation registered via `@serialize.register(type)`.

### Q23. Multiple Inheritance with Property Conflicts
Create two parent classes that both define a `@property` with the same name `value`, returning different things. Create a child class inheriting from both. Show which property wins (MRO), and how you'd explicitly call the "losing" parent's property from the child if needed.

### Q24. `__class_getitem__` — Generic-Style Classes
Implement `__class_getitem__` on a custom `TypedList` class so it supports the syntax `TypedList[int]` and `TypedList[str]` (like `list[int]` in Python 3.9+). The returned object should carry the type information and enforce it when items are added.

### Q25. Challenge — Plugin Architecture with Abstract Base and Registry
Design a complete plugin system:
1. An abstract base class `ReportGenerator` with abstract methods `generate(data)` and a class property `format_name`
2. A `ReportRegistry` that auto-registers all subclasses using `__init_subclass__`
3. Three concrete implementations: `PDFReport`, `CSVReport`, `HTMLReport` (simulated — just print what they'd do)
4. A factory function `get_generator(format_name)` that looks up the registry and returns the right class
5. A runner that takes a list of format names and data, generates all reports, catching `NotImplementedError` for unsupported formats gracefully
Demonstrate the full pipeline.
